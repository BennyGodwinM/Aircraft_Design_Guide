# 2.3 — Requirements Engineering

Writing requirements looks deceptively simple. In practice it is one of the more
consequential technical activities on any aircraft programme. A well-formed
requirements set enables efficient design, clear verification planning, and
traceable change management. A poorly formed set generates scope ambiguity, missed
interfaces, and disputes during test and certification that cost far more to resolve
than they would have cost to prevent.

This section covers the structure of a requirements set, the characteristics of a
well-written requirement, and the tools used to manage requirements through the
programme lifecycle.

## Types of Requirements

Aircraft programme requirements fall into several categories, each serving a
different analytical purpose.

**Functional requirements** define what the system must do: "The aircraft shall be
capable of transporting a minimum of 180 passengers with standard carry-on baggage."
They specify capability without constraining the mechanism.

**Performance requirements** attach measurable levels to functional capabilities:
"The aircraft shall achieve a balanced field length of 2,400 m or less at MTOW, ISA,
sea level." They quantify the functional requirement.

**Interface requirements** define how the aircraft communicates or physically
connects with external systems: "The aircraft fuel system shall accept single-point
pressure refuelling through an STPA NATO standard coupling in compliance with
MIL-PRF-26488." They govern the boundaries between the aircraft and its operating
environment.

**Physical requirements** constrain geometry and mass: "The aircraft wingspan shall
not exceed 64.0 m to maintain Code E airport compatibility." They translate
operational constraints into geometric bounds.

**Environmental requirements** specify the conditions the aircraft must operate
within and survive: "All control surface actuators shall be qualified for continuous
operation at ambient temperatures from −55 °C to +70 °C."

**Safety requirements** mandate survivability and failure behaviour: "No single
failure shall result in loss of all flight control authority in any axis." They
implement the failure tolerance philosophy required by the certification basis.

The boundary between categories is often blurry. A requirement such as "Maximum
demonstrated crosswind component ≥ 35 knots" is simultaneously a performance
requirement and a safety requirement. The category matters less than the content:
what matters is that every requirement is unambiguous, measurable, and traceable
{cite}`nasa2017seh`.

## What Makes a Requirement Good

The SMART framework — Specific, Measurable, Achievable, Relevant, and Testable — is
the standard quality checklist for individual requirements {cite}`incose2015`. In
aerospace, testability is the single most important criterion. A requirement that
cannot be verified by test, analysis, inspection, or demonstration should not be in
the document. "Passengers shall find the cabin comfortable" is an aspiration. "Cabin
noise level at the passenger seat shall be ≤ 72 dB(A) during cruise at FL350" is a
requirement: specific, measurable, and testable with a calibrated sound level meter.

Common failure modes in aerospace requirements writing:

*Ambiguity* — the requirement can be interpreted in more than one way. "The aircraft
shall have adequate fuel reserves" is ambiguous. "The aircraft shall carry sufficient
fuel to complete the design mission plus 45 minutes of final reserve fuel at 1,500 ft
at maximum continuous power" is not.

*Solution prescription* — the requirement specifies a design choice rather than a
capability. "The wing shall use a supercritical aerofoil section" is a design
decision, not a requirement. "The aircraft shall achieve cruise drag CD ≤ 0.024 at
CL = 0.52 at FL350" is a requirement that allows the designer to choose any aerofoil
that achieves the target.

*Infeasibility* — the requirement cannot be met within the physical or economic
constraints of the programme. Identifying an infeasible requirements set early,
before programme commitment, is one of the highest-value contributions a conceptual
designer can make. Infeasibility discovered during detailed design or test costs
orders of magnitude more to resolve.

*Unmeasurability* — the requirement cannot be verified by any available method. "The
aircraft shall have excellent handling qualities" fails here. "The aircraft shall
demonstrate Level 1 handling qualities per MIL-HDBK-1797 throughout the normal
flight envelope" does not.

## The Requirements Hierarchy and the V-Model

Requirements are organised in a hierarchy that flows from the top-level programme
need through successive levels of system decomposition {cite}`nasa2017seh`. The
**Mission Need Statement** (or, in defence programmes, the Initial Capabilities
Document) establishes the capability gap at the top. System-level requirements
describe what the complete aircraft must do. Subsystem-level requirements allocate
system requirements to major assemblies. Component-level requirements specify
individual parts or assemblies.

This hierarchy is not just an organisational tool — it is the mechanism by which
design responsibility is decomposed and allocated to development teams. The V-model
formalises the relationship between decomposition on the left arm and verification
on the right arm, linking each requirements level to its corresponding test or
analysis activity.

Figure 1 shows the V-model with the requirements hierarchy on the left arm and the
verification and validation activities on the right. Horizontal arrows connect each
decomposition level to its corresponding verification level, showing that system
requirements are verified at system integration test, subsystem requirements are
verified at subsystem test, and component requirements are verified at component
test or analysis.


````python
import matplotlib as mpl
mpl.rcParams.update({
    'figure.facecolor': 'none', 'axes.facecolor': 'none',
    'savefig.facecolor': 'none', 'savefig.transparent': True,
    'axes.edgecolor': '#888888', 'axes.labelcolor': '#cccccc',
    'text.color': '#cccccc', 'xtick.color': '#cccccc', 'ytick.color': '#cccccc',
    'grid.color': '#555555', 'font.size': 14, 'axes.titlesize': 16,
    'axes.labelsize': 14, 'xtick.labelsize': 13, 'ytick.labelsize': 13,
    'legend.fontsize': 13, 'legend.title_fontsize': 13,
    'figure.titlesize': 16, 'lines.linewidth': 2.0,
})
OKI = ['#0072B2','#D55E00','#009E73','#E69F00','#56B4E9','#CC79A7','#F0E442','#333333']
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

fig, ax = plt.subplots(figsize=(12, 7))
fig.patch.set_facecolor('white'); ax.set_facecolor('white')
ax.set_xlim(0, 12); ax.set_ylim(0, 8); ax.axis('off')

# Left arm boxes (decomposition): (x_centre, y_centre, label_top, label_bot)
left_boxes = [
    (1.5, 7.0, 'Mission Need', 'Statement / CONOPS'),
    (2.5, 5.5, 'System-Level', 'Requirements (L1)'),
    (3.5, 4.0, 'Subsystem', 'Requirements (L2)'),
    (4.5, 2.5, 'Component', 'Requirements (L3)'),
]
right_boxes = [
    (10.5, 7.0, 'Operational Test', '& Evaluation'),
    ( 9.5, 5.5, 'System Integration', '& Test'),
    ( 8.5, 4.0, 'Subsystem', 'Integration & Test'),
    ( 7.5, 2.5, 'Component Test', '& Analysis'),
]
bottom_box = (6.0, 1.0, 'Build, Code,', 'Fabricate')

col_l = [OKI[1], OKI[0], OKI[2], OKI[3]]
col_r = [OKI[1], OKI[0], OKI[2], OKI[3]]

BW, BH = 2.2, 0.9

def draw_box(ax, cx, cy, t1, t2, col, bw=BW, bh=BH):
    r = mpatches.FancyBboxPatch((cx-bw/2, cy-bh/2), bw, bh,
        boxstyle='round,pad=0.06', facecolor=col, edgecolor='white', lw=1.5, zorder=3)
    ax.add_patch(r)
    ax.text(cx, cy+0.17, t1, ha='center', va='center', fontsize=9, fontweight='bold',
            color='white', zorder=4)
    ax.text(cx, cy-0.20, t2, ha='center', va='center', fontsize=8.5, color='white',
            alpha=0.90, zorder=4)

for (cx,cy,t1,t2), col in zip(left_boxes, col_l):
    draw_box(ax, cx, cy, t1, t2, col)
for (cx,cy,t1,t2), col in zip(right_boxes, col_r):
    draw_box(ax, cx, cy, t1, t2, col)
draw_box(ax, *bottom_box, OKI[5], bw=2.4, bh=0.9)

# V-shape arrows (left arm down, right arm up)
lx = [b[0] for b in left_boxes]; ly = [b[1] for b in left_boxes]
rx = [b[0] for b in right_boxes]; ry = [b[1] for b in right_boxes]
bx, by = bottom_box[0], bottom_box[1]

# Connect left arm
for i in range(len(lx)-1):
    ax.annotate('', xy=(lx[i+1], ly[i+1]+BH/2), xytext=(lx[i], ly[i]-BH/2),
        arrowprops=dict(arrowstyle='->', color='#555555', lw=1.8))
ax.annotate('', xy=(bx-BW/2, by+BH/2), xytext=(lx[-1], ly[-1]-BH/2),
    arrowprops=dict(arrowstyle='->', color='#555555', lw=1.8))
# Connect right arm
ax.annotate('', xy=(rx[-1], ry[-1]-BH/2), xytext=(bx+BW/2, by+BH/2),
    arrowprops=dict(arrowstyle='->', color='#009E73', lw=1.8))
for i in range(len(rx)-1, 0, -1):
    ax.annotate('', xy=(rx[i-1], ry[i-1]-BH/2), xytext=(rx[i], ry[i]+BH/2),
        arrowprops=dict(arrowstyle='->', color='#009E73', lw=1.8))

# Horizontal verification arrows
for (lx_,ly_), (rx_,ry_) in zip(
        [(lx[i],ly[i]) for i in range(len(lx))],
        [(rx[i],ry[i]) for i in range(len(rx))]):
    ax.annotate('', xy=(rx_-BW/2, ry_), xytext=(lx_+BW/2, ly_),
        arrowprops=dict(arrowstyle='->', color='#aaaaaa', lw=1.2, ls='dashed'))

ax.text(1.0, 3.0, 'Decomposition\n(Design)', ha='center', va='center',
        fontsize=9.5, color='#555555', rotation=70)
ax.text(11.2, 3.0, 'Verification\n& Validation', ha='center', va='center',
        fontsize=9.5, color='#009E73', rotation=-70)
ax.set_title('V-Model Systems Engineering Lifecycle', fontsize=13,
             fontweight='bold', color='#111111')
plt.tight_layout(); plt.show()
````

*Figure 1 — The V-model, showing the relationship between requirements
decomposition (left arm) and verification and validation (right arm). Horizontal
arrows link each requirements level to its corresponding test or analysis activity.
The model is iterative: problems discovered on the right arm require returning to
the corresponding level on the left arm for requirements revision. Adapted from
{cite}`nasa2017seh` and {cite}`incose2015`.*

## Requirements Traceability Matrix

The requirements traceability matrix (RTM) is the formal record of the relationship
between requirements at each level of the hierarchy and between requirements and
their verification evidence. At minimum, an RTM records for each requirement: its
unique identifier, its parent requirement (the higher-level requirement it is derived
from), the design element responsible for meeting it, and the verification method
and reference.

Bidirectional traceability means every requirement can be traced forward to its
implementing design element and backward to its parent. Forward traceability
confirms every requirement has been addressed in the design. Backward traceability
confirms every design decision is justified by a requirement — that there is no
design element without a requirement and no scope that was not formally agreed.

Scope creep — the uncontrolled expansion of programme scope that drives cost and
schedule growth — almost always enters through requirements that are not in the RTM.
A feature added because an engineer thought it was a good idea, without a formal
requirement backing it, cannot be traced. An RTM maintained throughout the programme
is the single most effective tool for detecting scope creep early, when correction is
cheap {cite}`incose2015`.

Figure 2 shows a schematic RTM, illustrating how requirements trace to implementing
subsystems and to verification methods.


````python
import matplotlib as mpl
mpl.rcParams.update({
    'figure.facecolor': 'none', 'axes.facecolor': 'none',
    'savefig.facecolor': 'none', 'savefig.transparent': True,
    'axes.edgecolor': '#888888', 'axes.labelcolor': '#cccccc',
    'text.color': '#cccccc', 'xtick.color': '#cccccc', 'ytick.color': '#cccccc',
    'grid.color': '#555555', 'font.size': 14, 'axes.titlesize': 16,
    'axes.labelsize': 14, 'xtick.labelsize': 13, 'ytick.labelsize': 13,
    'legend.fontsize': 13, 'legend.title_fontsize': 13,
    'figure.titlesize': 16, 'lines.linewidth': 2.0,
})
OKI = ['#0072B2','#D55E00','#009E73','#E69F00','#56B4E9','#CC79A7','#F0E442','#333333']
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

reqs = [
    'SYS-REQ-001 Balanced field length ≤ 2 400 m',
    'SYS-REQ-002 Cruise Mach ≥ 0.82 at FL350',
    'SYS-REQ-003 MTOW ≤ 79 000 kg',
    'AERO-REQ-001 CLmax (landing) ≥ 2.65',
    'PROP-REQ-001 Net thrust ≥ 120 kN (ISA, SL)',
    'STR-REQ-001  Ultimate load factor = 1.5 × limit',
    'SYS-REQ-004 Cabin noise ≤ 72 dB(A) cruise',
    'SYS-REQ-005 Single-pt refuel ≤ 12 min',
]
subsystems = ['Aerodyn.', 'Structures', 'Propulsion', 'Systems', 'Verification\nMethod']
# (row, col) = 1 means primary, 0.5 = contributing, 0 = N/A
# Cols: Aero, Struct, Prop, Sys, Method
matrix = [
    [1,   0.5,  0.5,  0,   0],   # field length
    [1,   0,    0.5,  0,   0],   # cruise speed
    [0.5, 1,    0.5,  0.5, 0],   # MTOW
    [1,   0,    0,    0,   0],   # CLmax
    [0,   0,    1,    0,   0],   # thrust
    [0,   1,    0,    0,   0],   # load factor
    [0.5, 0.5,  0,    1,   0],   # cabin noise
    [0,   0,    0,    1,   0],   # refuel time
]
methods = ['FT','FT','W+A','FT','GT','A','FT','FT']
for i, m in enumerate(methods):
    matrix[i][4] = m  # string in last col

fig, ax = plt.subplots(figsize=(11, 5.5))
fig.patch.set_facecolor('white'); ax.set_facecolor('white')
nR, nC = len(reqs), len(subsystems)

for r in range(nR):
    for c in range(nC-1):
        val = matrix[r][c]
        if val == 1:
            col = OKI[0]; alpha = 0.85
        elif val == 0.5:
            col = OKI[4]; alpha = 0.55
        else:
            col = '#f0f0f0'; alpha = 1.0
        rect = mpatches.Rectangle((c, nR-r-1), 1, 1, facecolor=col,
                                   edgecolor='white', lw=1.5, alpha=alpha, zorder=2)
        ax.add_patch(rect)
        if val == 1:
            ax.text(c+0.5, nR-r-0.5, 'P', ha='center', va='center',
                    fontsize=9, fontweight='bold', color='white', zorder=3)
        elif val == 0.5:
            ax.text(c+0.5, nR-r-0.5, 'C', ha='center', va='center',
                    fontsize=9, color='white', zorder=3)
    # Method column
    ax.text(nC-0.5, nR-r-0.5, methods[r], ha='center', va='center',
            fontsize=9, color='#333333', zorder=3,
            bbox=dict(boxstyle='round,pad=0.2', facecolor='#e8f4e8', edgecolor='#aaaaaa'))

ax.set_xlim(0, nC); ax.set_ylim(0, nR)
ax.set_xticks(np.arange(nC)+0.5); ax.set_xticklabels(subsystems, fontsize=9.5, color='#222222')
ax.set_yticks(np.arange(nR)+0.5)
ax.set_yticklabels([r[:38] for r in reqs[::-1]], fontsize=8.5, color='#222222')
ax.tick_params(axis='both', length=0)
for sp in ax.spines.values(): sp.set_visible(False)
ax.set_title('Schematic Requirements Traceability Matrix (RTM)\n'
             'P = primary responsible subsystem   C = contributing subsystem   '
             'Method: FT = flight test, A = analysis, W = weighing, GT = ground test',
             fontsize=9.5, color='#333333')
plt.tight_layout(); plt.show()
````

*Figure 2 — A schematic RTM for a commercial transport aircraft programme.
Blue cells (P) indicate the subsystem with primary responsibility for satisfying
the requirement. Lighter blue cells (C) indicate contributing subsystems. The
rightmost column records the planned verification method. A real RTM contains
hundreds or thousands of rows; this figure illustrates structure, not scale.
Constructed from methodology in {cite}`nasa2017seh` and {cite}`incose2015`.*

## Model-Based Systems Engineering

Traditional systems engineering manages requirements in documents: Word files, PDFs,
or dedicated tools such as IBM DOORS or Jama Connect. Model-Based Systems Engineering
(MBSE) replaces document-centric management with a formal model, expressed in SysML
(Systems Modelling Language), that captures requirements, their relationships, the
system architecture, and allocated functions in a single coherent database
{cite}`nasa2017seh`.

The advantages of MBSE are significant for complex aircraft programmes. Requirements
are stored with their relationships explicit and machine-readable, rather than
implied by document structure. Impact analysis — tracing the consequences of a
proposed change through the full system — can be automated. Interfaces between
subsystems are formally defined in the model, reducing the chance of mismatches
reaching hardware.

MBSE does not eliminate the need for clear, well-formed individual requirements. A
poorly written requirement is just as ambiguous in a SysML Requirements Diagram as
it is in a Word document.

## Regulatory Standards for Requirements

**MIL-STD-961** specifies the format and content standards for defence specification
documents, including requirements documents. It defines how requirements shall be
structured, numbered, and presented in US military procurement contracts {cite}`milstd961`.

**DO-178C** (Software Considerations in Airborne Systems and Equipment Certification)
governs software requirements in certified avionics. It defines five Software Levels
(A through E) based on failure severity and specifies the requirements capture,
verification, and validation activities required at each level. For Level A — loss of
the aircraft — every software requirement must trace to a system requirement, and
every system requirement allocated to software must have complete, unambiguous,
verifiable software requirements {cite}`do178c`.

**DO-254** (Design Assurance Guidance for Airborne Electronic Hardware) provides
equivalent guidance for programmable logic devices used in avionics. Like DO-178C,
it defines hardware design assurance levels and requires traceability from high-level
hardware requirements through detailed design to implementation {cite}`do254`.

Both standards apply the same fundamental principle: requirements must be traceable,
verifiable, and managed under configuration control throughout the design life.

```{tip}
Number your requirements from Day 1. A requirement without a unique identifier
cannot be traced, cannot be tracked in an RTM, and cannot be formally changed. Use
a structured scheme: SYS-REQ-001 for system-level requirements, AERO-REQ-001 for
aerodynamics, PROP-REQ-001 for propulsion. Any text in a requirements document that
does not carry a number is commentary, not a requirement.
```

```{note}
DO-178C and DO-254 are published by RTCA, Inc. and adopted in the US by FAA
Advisory Circulars AC 20-115 and AC 20-152. EASA accepts the RTCA standards through
EUROCAE ED-12C and ED-80, which are technically equivalent. CAAC (China) has adopted
equivalent standards through CCAR-21 and CTSO certification.
```
