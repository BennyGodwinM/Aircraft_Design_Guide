# 2.7 — The Design Review Cycle

A design review is a formal, structured evaluation of the design at a defined
programme milestone, conducted by a review board to assess whether the design is
ready to proceed to the next phase. Reviews are the programme's primary mechanism
for controlling technical risk, identifying open issues before they become expensive,
and obtaining formal stakeholder approval to proceed.

## The Cost of Late Change

The most important concept in understanding why design reviews matter is the
cost-of-change curve. The cost of correcting a design decision is approximately an
order of magnitude higher at each successive programme phase than at the previous
one {cite}`blanchard2011`.

A requirements error discovered during the requirements review (SRR) costs a few
days of discussion and document revision. The same error discovered during
preliminary design may require weeks of re-analysis across multiple subsystems.
Discovered at critical design review, it may require scrapping tooling already
committed. Discovered during first article assembly, it requires rework of delivered
hardware. Discovered in service, it generates an Airworthiness Directive affecting
every aircraft in the fleet.

This cost scaling is the reason programmes invest disproportionate effort in
front-end reviews. The review process is not bureaucratic overhead — it is the
engineering programme's primary cost-control tool.

## The Standard Review Cycle

Most civil and military aircraft programmes follow a review cycle derived from the
US Department of Defense acquisition process {cite}`incose2015`. The reviews are:

**System Requirements Review (SRR)** occurs at the end of the concept phase, before
preliminary design begins. The SRR assesses whether the system requirements — TLARs
and Level 1 requirements — are complete, consistent, and feasible. The review board
examines whether every stakeholder need has been captured, whether the requirements
are testable, whether they are achievable within programme resources, and whether any
requirements conflict. A programme that passes SRR has a documented, agreed
requirements baseline. A programme that enters preliminary design without passing SRR
is committing resources to a design that may need to be substantially revised when
requirements are eventually agreed.

**Preliminary Design Review (PDR)** occurs at the end of preliminary design, after
the baseline configuration has been selected and subsystem requirements allocated.
PDR assesses whether the baseline configuration is credible — whether the top-level
performance analysis shows the design closes on the requirements, whether subsystem
requirements are complete and allocated, whether major interfaces have been defined
in ICDs, and whether technical risks are understood and managed. PDR is the last
review at which substantial configuration changes can be made without major programme
impact.

**Critical Design Review (CDR)** occurs at the end of detail design, before
manufacturing release. CDR assesses whether the detailed design meets all
requirements at the component and subsystem level, whether the design is producible
within the agreed manufacturing process and tooling base, and whether the
verification plan covers all requirements. Passing CDR means the design is frozen:
drawings can be released to manufacturing, tooling committed, and suppliers placed
on long-lead procurement. Reopening CDR items after CDR is the single most expensive
category of late design change {cite}`nasa2017seh`.

**Test Readiness Review (TRR)** occurs before the start of each major test phase —
before ground integration testing and again before first flight. TRR assesses whether
the test article is compliant with the CDR design baseline, whether test procedures
are approved, whether instrumentation is calibrated and functional, and whether the
safety plan for the test is adequate. No test may begin before the TRR has formally
cleared it.

**Flight Readiness Review (FRR)** is a specific instantiation of TRR applied to the
first flight of a new aircraft or major variant. FRR is a full-programme review:
every open issue, every waiver to the as-built configuration, every known limitation
in the envelope cleared for the first flight, and every hazard associated with the
first flight test is reviewed by a board that includes programme management, the
certifying authority representative, the chief test pilot, and the head of the flight
test instrumentation team.

entry criteria, open-item counts, and design freeze dates for each review.


````python
import matplotlib as mpl
mpl.rcParams.update({
    'figure.facecolor': 'none', 'axes.facecolor': 'none',
    'savefig.facecolor': 'none', 'savefig.transparent': True,
    'axes.edgecolor': '#888888', 'axes.labelcolor': '#cccccc',
    'text.color': '#cccccc', 'xtick.color': '#cccccc', 'ytick.color': '#cccccc',
    'grid.color': '#555555', 'font.size': 14, 'axes.titlesize': 16,
    'axes.labelsize': 14, 'xtick.labelsize': 13, 'ytick.labelsize': 13,
    'legend.fontsize': 13, 'legend.title_fontsize': 13,
    'figure.titlesize': 16, 'lines.linewidth': 2.0,
})
OKI = ['#0072B2','#D55E00','#009E73','#E69F00','#56B4E9','#CC79A7','#F0E442','#333333']
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 7),
    gridspec_kw={'height_ratios': [2, 1]}, sharex=True)
fig.patch.set_facecolor('white')
ax1.set_facecolor('#fafafa'); ax2.set_facecolor('#fafafa')

# Timeline in arbitrary programme months (0 = programme start)
reviews = [
    ('SRR', 6,   OKI[2]),
    ('PDR', 18,  OKI[0]),
    ('CDR', 36,  OKI[3]),
    ('TRR', 50,  OKI[1]),
    ('FRR', 54,  OKI[1]),
]
phases_gantt = [
    ('Concept / Requirements', 0, 8,  OKI[4], 3.2),
    ('Preliminary Design',     7, 13, OKI[0], 2.5),
    ('Detail Design',          19, 19, OKI[2], 1.8),
    ('Manufacture & Build',    37, 14, OKI[3], 1.1),
    ('Integration & Test',     50, 14, OKI[1], 0.4),
]

for lbl, start, dur, col, y in phases_gantt:
    r = mpatches.FancyBboxPatch((start, y-0.28), dur, 0.56,
        boxstyle='round,pad=0.04', facecolor=col, edgecolor='white', lw=1.2,
        zorder=3, alpha=0.80)
    ax1.add_patch(r)
    ax1.text(start+dur/2, y, lbl, ha='center', va='center', fontsize=8.5,
             fontweight='bold', color='white', zorder=4)

for lbl, t, col in reviews:
    ax1.axvline(t, color=col, lw=2.2, ls='--', alpha=0.9, zorder=5)
    ax1.text(t, 4.0, lbl, ha='center', fontsize=10, fontweight='bold', color=col)

ax1.set_ylim(0, 4.5); ax1.set_yticks([])
for sp in ax1.spines.values(): sp.set_edgecolor('#cccccc')
ax1.set_title('Design Review Cycle — Programme Timeline and Open-Item Evolution',
              fontsize=12, fontweight='bold', color='#111111')

# Open items curve (schematic)
t_arr = np.linspace(0, 64, 300)
open_items = (
    80 * np.exp(-((t_arr-6)/3)**2)      # SRR spike then close
  + 200 * np.exp(-((t_arr-18)/4)**2)    # PDR spike
  + 350 * np.exp(-((t_arr-36)/5)**2)    # CDR spike
  + 150 * np.exp(-((t_arr-50)/3)**2)    # TRR spike
  + 50  * np.exp(-((t_arr-54)/2)**2)    # FRR spike
  + 20
)
ax2.fill_between(t_arr, open_items, alpha=0.15, color=OKI[1])
ax2.plot(t_arr, open_items, color=OKI[1], lw=2.2)
for lbl, t, col in reviews:
    ax2.axvline(t, color=col, lw=2.2, ls='--', alpha=0.7, zorder=5)
ax2.set_xlim(0, 64); ax2.set_ylim(0, 420)
ax2.set_xlabel('Programme Month', fontsize=10, color='#333333')
ax2.set_ylabel('Open Review Items', fontsize=10, color='#333333')
ax2.tick_params(colors='#444444')
ax2.grid(color='#e0e0e0', lw=0.8)
for sp in ax2.spines.values(): sp.set_edgecolor('#cccccc')

plt.tight_layout(); plt.show()
````

*Figure 2 — Design review cycle mapped to a representative programme timeline. Upper

A rising open-item count between reviews is a warning sign of programme difficulty.
Based on programme structure in {cite}`nasa2017seh` and {cite}`incose2015`.*

## Entry and Exit Criteria

Each review has defined entry criteria — the minimum state of completion the design
and documentation must be in before the review begins — and exit criteria — the
conditions that must be satisfied before the review board issues a "proceed"
recommendation. The entry criteria are checked by the programme's systems engineering
function before the review date; a review that begins without meeting entry criteria
is almost always forced to adjourn for re-planning, which costs schedule.

Exit criteria for SRR typically include: all Level 1 requirements documented with
unique identifiers, all requirements reviewed for testability, the RTM populated for
Level 1, no open requirements conflicts, and the programme risk register updated with
all critical risks assessed.

Exit criteria for CDR typically include: all drawings released at the revision
required for manufacturing, all Level 2 and Level 3 requirements verified by analysis
or shown to be verifiable by test, all critical waivers and deviations approved,
supplier first article inspection plans approved, and the flight test programme
baseline established.

Tracking open items from each review — issues raised by the review board that have
not yet been resolved — is the programme office's primary tool for monitoring design
maturity between reviews. An open items list growing between PDR and CDR indicates a
programme that is not converging; the trend is as important as the absolute count.

## Decision Authority at Each Review

Design reviews in government and military programmes carry formal decision authority.
The review board for an SRR is chaired at a level appropriate to the programme stage
— typically a senior programme manager. PDR and CDR boards are typically chaired by
the programme's technical authority, with representation from the customer's programme
office. For major defence programmes, CDR may be chaired at the level of the Service
Acquisition Executive {cite}`nasa2017seh`.

In commercial aviation, the equivalent structure is established through the Type
Inspection Authorisation and the Organisation Approval process. The FAA or EASA
delegates review authority to the manufacturer under Design Organisation Approval
for established manufacturers, while retaining direct involvement for novel
configurations, new entrant manufacturers, or specific critical systems.

The practical implication is that design reviews are not internal engineering events.
They are milestones in the regulatory and contractual record of the programme that
include formal stakeholder representation, produce binding decisions, and generate
documentary records that form part of the certification data package. Treating a
design review as an internal team meeting with an agenda is a category error.

```{tip}
Prepare for design reviews by working backward from the exit criteria. Write the
exit criteria document before preliminary design starts. Every week of preliminary
design should be moving you toward satisfying those criteria, not toward filling up
presentation slides. If you cannot describe how a design week advances your readiness
for PDR, you are working on the wrong problem.
```

```{note}
The formal review cycle described here applies primarily to civil and military
aircraft programmes in the US and Europe. The UK MOD uses the CADMID (Concept,
Assessment, Demonstration, Manufacture, In-service, Disposal) framework with
equivalent review gates. ESA uses a phase/milestone model aligned to NASA's. COMAC
in China follows CAAC certification requirements with review cycles aligned to type
certificate milestones.
```
