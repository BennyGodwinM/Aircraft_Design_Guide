# 2.6 — The Systems Engineering Process

Systems engineering is the discipline that manages the design of complex systems
through their full lifecycle, from the initial stakeholder need to retirement and
disposal. For aircraft — where hundreds of thousands of parts must work together
reliably in a demanding environment — no uncoordinated collection of discipline
experts can produce a safe, certifiable, manufacturable, maintainable result.
Someone has to be responsible for making sure all the pieces fit together, in every
sense {cite}`nasa2017seh`.

## Sequential Versus Concurrent Engineering

The traditional aircraft design process was largely sequential: aerodynamics defined
the external geometry, structures designed the internal load path, systems engineers
routed their lines through whatever space was left, and manufacturing figured out how
to build the result. Each discipline worked in relative isolation until it was
"done" and handed off to the next.

This model has a consistent failure mode. When manufacturing discovers that the
aerodynamic design requires a machined compound-curved panel that cannot be produced
within cost or schedule, the problem goes back to aerodynamics. By that point,
structural design has been completed around the panel geometry. Changing the panel
means re-analysing the structure. The cascade of changes grows as it propagates, and
the cost of fixing the problem at the manufacturing stage is orders of magnitude
higher than it would have been if manufacturing had been consulted during the initial
aerodynamic configuration trade {cite}`raymer2018`.

Concurrent engineering replaces the sequential handoff with simultaneous, coordinated
analysis by all relevant disciplines from the earliest design stage. Aerodynamics,
structures, manufacturing, systems, and supportability are all represented when
geometry decisions are being made, and their constraints are all visible during
conceptual configuration trade studies.

This is not a management preference — it has a documented effect on programme
outcomes. Programmes that implemented structured concurrent engineering practices
in the 1980s and 1990s, particularly on US defence programmes, showed consistent
reductions in late-stage design changes and in production cost relative to
sequential-process programmes of equivalent complexity {cite}`blanchard2011`.

## Integrated Product Teams

The organisational mechanism for concurrent engineering is the Integrated Product
Team (IPT). An IPT is a cross-functional team co-located and jointly responsible
for a specific product or major assembly. Rather than organising the design around
engineering disciplines — an aerodynamics department, a structures department — the
IPT structure organises around product breakdown: a wing IPT, a fuselage IPT, a
propulsion IPT {cite}`nasa2017seh`.

Each IPT includes representatives from all disciplines relevant to its product. The
wing IPT includes aerodynamicists, structural designers, systems engineers
responsible for fuel and hydraulic routing, manufacturing engineers responsible for
wing assembly, and representatives from the maintenance and supportability function.
All of these representatives are simultaneously engaged in the design, and the
constraints of each discipline are visible to all others.

The IPT structure requires a Lead Systems Integrator (LSI) function that sits above
the IPT level to manage interfaces between IPTs. The wing IPT and the fuselage IPT
both have design responsibility at the wing-fuselage joint. Without a formal
Interface Control Document (ICD) managed by the LSI, the two IPTs will make
incompatible assumptions about the joint, and the incompatibility will be discovered
when the wing and fuselage are brought together for first join. At that stage, fixing
it is extremely expensive.

## The Systems Engineering Lifecycle

The systems engineering process follows a lifecycle that spans from the initial
concept through disposal. The NASA and INCOSE lifecycle models share the same
fundamental structure {cite}`incose2015`:

*Pre-Phase A — Concept Studies*: exploratory work to define stakeholder needs,
identify the mission space, and bound the design problem. The output is an initial
concept with sufficient definition to support a programme initiation decision.

*Phase A — Concept Development*: the formal conceptual design phase. TLARs are
defined, candidate configurations are sized parametrically, and a preferred concept
is selected. This phase ends with the System Requirements Review (SRR).

*Phase B — Preliminary Design*: the baseline configuration is defined in sufficient
detail for subsystem design to begin. Requirements are allocated to subsystems,
interface control documents are written, and the design is analysed to validate
compliance with system-level requirements. This phase ends with the Preliminary
Design Review (PDR).

*Phase C — Detail Design*: individual components are designed, analysed, and
documented. Tooling and manufacturing processes are defined. Supplier contracts are
placed. This phase ends with the Critical Design Review (CDR), which certifies the
design is ready for first build.

*Phase D — System Assembly, Integration, and Test*: hardware is built, integrated,
and tested against the requirements established in Phases A through C. This phase
includes first flight and the certification flight test programme.

*Phase E — Operations and Sustainment*: the aircraft is in service. Modifications,
airworthiness directives, and life extensions are managed. Reliability and
maintainability data from service experience feed back into requirements for
derivative programmes.

*Phase F — Closeout*: retirement, disposal, and lessons-learned capture.

The lifecycle is iterative: programmes do not execute it strictly in sequence. A
requirement problem discovered in Phase D may require returning to Phase B for
requirements revision. Managing that iteration without uncontrolled scope expansion
is the central challenge of systems engineering practice.

Figure 2 shows the NASA/INCOSE lifecycle phases mapped to the key design reviews

````python
import matplotlib as mpl
mpl.rcParams.update({
    'figure.facecolor': 'none', 'axes.facecolor': 'none',
    'savefig.facecolor': 'none', 'savefig.transparent': True,
    'axes.edgecolor': '#888888', 'axes.labelcolor': '#cccccc',
    'text.color': '#cccccc', 'xtick.color': '#cccccc', 'ytick.color': '#cccccc',
    'grid.color': '#555555', 'font.size': 14, 'axes.titlesize': 16,
    'axes.labelsize': 14, 'xtick.labelsize': 13, 'ytick.labelsize': 13,
    'legend.fontsize': 13, 'legend.title_fontsize': 13,
    'figure.titlesize': 16, 'lines.linewidth': 2.0,
})
OKI = ['#0072B2','#D55E00','#009E73','#E69F00','#56B4E9','#CC79A7','#F0E442','#333333']
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

fig, ax = plt.subplots(figsize=(12, 5))
fig.patch.set_facecolor('white'); ax.set_facecolor('#fafafa')

phases = ['Pre-A\nConcept\nStudies', 'Phase A\nConcept\nDev.', 'Phase B\nPrelim.\nDesign',
          'Phase C\nDetail\nDesign', 'Phase D\nInt. &\nTest', 'Phase E\nOps &\nSustain']
phase_cols = [OKI[4], OKI[0], OKI[2], OKI[3], OKI[1], OKI[5]]
x_starts = [0, 1.2, 2.6, 4.1, 5.7, 7.8]
x_widths = [1.1, 1.3, 1.4, 1.5, 2.0, 3.2]

for x, w, lbl, col in zip(x_starts, x_widths, phases, phase_cols):
    r = mpatches.FancyBboxPatch((x+0.05, 3.5), w-0.1, 1.2,
        boxstyle='round,pad=0.06', facecolor=col, edgecolor='white', lw=1.5, zorder=3, alpha=0.85)
    ax.add_patch(r)
    ax.text(x+w/2, 4.1, lbl, ha='center', va='center', fontsize=8.5,
            fontweight='bold', color='white', multialignment='center', zorder=4)

# Design reviews
reviews = [('SRR', 1.2), ('PDR', 2.6), ('CDR', 4.1), ('TRR', 5.7), ('FRR', 6.2)]
for lbl, x in reviews:
    ax.plot([x, x], [3.3, 3.5], color='#D55E00', lw=2.0, zorder=5)
    ax.scatter(x, 3.3, s=80, color='#D55E00', zorder=6, edgecolors='white', lw=1.5)
    ax.text(x, 3.05, lbl, ha='center', fontsize=9.5, fontweight='bold', color='#D55E00')

# Maturity curves
x_full = np.linspace(0, 11, 300)

def sigmoid(x, x0, k): return 1/(1+np.exp(-k*(x-x0)))

req_maty  = sigmoid(x_full, 2.0, 1.8)
des_maty  = sigmoid(x_full, 3.8, 1.5)
verif_mat = sigmoid(x_full, 5.5, 1.8)

ax.plot(x_full, req_maty*2.5+0.2, color=OKI[0], lw=2.2, label='Requirements definition')
ax.plot(x_full, des_maty*2.5+0.2, color=OKI[2], lw=2.2, label='Design maturity')
ax.plot(x_full, verif_mat*2.5+0.2, color=OKI[3], lw=2.2, label='Verification completeness')

ax.set_xlim(0, 11); ax.set_ylim(0, 5.2)
ax.set_xticks([]); ax.set_yticks([0.2, 0.2+1.25, 0.2+2.5])
ax.set_yticklabels(['0%', '50%', '100%'], fontsize=9, color='#444444')
ax.set_ylabel('Programme Completeness', fontsize=10, color='#333333')
ax.set_title('Systems Engineering Lifecycle — Phase, Review, and Maturity Evolution',
             fontsize=12, fontweight='bold', color='#111111')
ax.legend(loc='center left', fontsize=9.5, framealpha=0.9)
ax.grid(axis='y', color='#e0e0e0', lw=0.8)
for sp in ax.spines.values(): sp.set_edgecolor('#cccccc')
plt.tight_layout(); plt.show()
````

*Figure 2 — The NASA/INCOSE systems engineering lifecycle phases (coloured bands)
with major design reviews (orange markers) and three maturity curves. Requirements
definition matures earliest — approaching 100% by Phase B/PDR. Design maturity
follows through Phase C/CDR. Verification completeness ramps steeply during Phase D
integration and test. A programme in good health shows all three curves tracking
this order; a programme where verification is attempted before design has matured
typically requires expensive rework. Based on the lifecycle model in
{cite}`nasa2017seh` and the phase definitions in {cite}`incose2015`.*

## Model-Based Systems Engineering in Practice

Traditional systems engineering manages requirements and interfaces in documents —
Word files, PDFs, or dedicated tools such as IBM DOORS or Jama Connect. Model-Based
Systems Engineering (MBSE) replaces document-centric management with a formal model,
expressed in SysML, that captures requirements, their relationships, the system
architecture, and allocated functions in a single coherent database {cite}`nasa2017seh`.

What MBSE offers complex programmes that document-based approaches cannot match is
automated impact analysis. When a customer requests a 5% increase in design range,
a properly constructed MBSE model can propagate the change through the requirements
hierarchy, identify every requirement affected, flag every interface that needs
revisiting, and identify every component whose weight budget depends on the changed
parameter. The same analysis in a document-based system requires manual cross-
referencing of dozens of documents — a process that takes weeks and reliably misses
some dependencies.

MBSE does not eliminate the need for clear, well-formed individual requirements. It
is a management and analysis tool, not a substitute for engineering judgment.

```{tip}
Interface control is the most common failure point in complex systems engineering.
Define your interfaces formally — with ICDs, not informal agreements — as early in
the programme as possible. The F-35's weight growth problem was substantially driven
by interface surprises discovered late, when the cost of changing the interface
geometry was prohibitive. Early ICD discipline is cheap; late interface rework is not.
```
