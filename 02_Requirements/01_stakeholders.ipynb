# 2.1 — Stakeholder Identification & Analysis

No aircraft is designed for nobody. The moment a programme begins, there is already
a set of people and organisations whose lives will be shaped by what gets built —
who will buy it, fly it, fix it, certify it, or simply endure its noise overhead.
Identifying that set, understanding how they rank relative to each other, and
managing their competing demands is the first concrete task on any new aircraft
programme.

The formal term is *stakeholders*: any individual, group, or organisation that can
affect or be affected by the design, development, deployment, or disposal of the
aircraft {cite}`nasa2017seh`. The definition is broader than it first appears. The
regulator who sets the certification standard is a stakeholder. So is the ground
crew member fuelling the aircraft at 4 a.m. So is the community under the approach
path whose residents did not choose to live near an airport but bear the acoustic
cost of someone else's design decisions.

Stakeholder analysis is not a soft management exercise bolted onto the outside of
real engineering. It is a prerequisite for writing requirements. You cannot produce
a sound requirements document without first knowing whose requirements you are
capturing. A requirements process that skips stakeholder analysis ends up capturing
the loudest voices in the room, which is rarely the same thing as capturing the most
important ones {cite}`ullman2009`.

## A Three-Tier Taxonomy

Stakeholders fall into three tiers, separated by how directly they interact with the
vehicle and how much formal authority they hold over the design.

**Primary stakeholders** interact directly with the aircraft and have formal
relationships with the programme. For a commercial transport this includes the
airline, the flight crew, cabin crew, passengers, line maintenance technicians who
turn the aircraft around between flights, and heavy-maintenance engineers who conduct
scheduled deep-strip overhauls. Each group has specific, verifiable needs that must
eventually become engineering requirements.

Maintainers deserve particular emphasis because they are the group most reliably
excluded from early stakeholder engagement, and their exclusion most reliably
produces expensive problems. The F-35 is the obvious example: engine removal on
certain variants requires partial disassembly of the forward airframe, a direct
consequence of access constraints not adequately addressed at the conceptual stage
{cite}`raymer2018`. On the F-16, the same task takes roughly four hours under field
conditions. That gap is the compounded result of decisions made when the programme
assumed maintainability could be addressed later. It cannot.

For a military strike aircraft, the primary tier includes the pilot, mission
commander, intelligence analysts who task the aircraft, ground crew, and the
depot-level maintainer working under a forward-deployed shelter. For a small
commercial UAV, it collapses to the remote pilot, payload operator, and logistics
team between sorties. The composition changes; the principle that all primary
stakeholders must be represented in the requirements process does not.

**Secondary stakeholders** do not interact with the aircraft directly but exercise
significant authority over what it can be. The certifying authority — FAA, EASA,
CAAC, or the relevant military body — sits here, as does ICAO whose standards
underpin national regulations. Airport authorities impose physical constraints:
wingspan determines gate compatibility, noise limits restrict approach profiles, and
pavement bearing strength sets maximum ramp weights. Air Traffic Control imposes
avionics and communication standards that are not negotiable. Major Tier-1 suppliers
— engine manufacturers, avionics integrators — are secondary stakeholders whose
technical commitments shape the design from outside {cite}`sadraey2012`.

**Tertiary stakeholders** are affected by the aircraft's operation without any formal
relationship with the programme. Communities under flight paths bear acoustic loads
they had no role in negotiating. Environmental bodies assess the emissions profile.
Insurers price the actuarial risk. In urban air mobility programmes this tier grows
substantially: eVTOL vehicles operating over dense city environments create a
tertiary stakeholder population — residents, city councils, emergency services,
competing ground transport — whose concerns have the political weight to constrain or
cancel a programme even after technical certification is achieved.

## The Power–Interest Matrix

Not every stakeholder warrants the same management intensity. The power–interest
matrix, introduced by Mendelow (1991) and widely adopted in systems engineering
practice {cite}`mendelow1991`, places stakeholders in a two-dimensional space and
maps them to four management strategies.

The horizontal axis is *interest*: how much does this stakeholder care about the
outcome? The vertical axis is *power*: how much authority does this stakeholder have
to influence design decisions, approve the programme, or block it? The four
quadrants produce four strategies:

*Manage Closely* (high power, high interest): These stakeholders carry the most
requirements weight. Their objections can stop the aircraft in its tracks. For a
commercial transport, this quadrant typically contains the certifying authority and
the launch customer.

*Keep Satisfied* (high power, low interest): These stakeholders have authority but do
not have strong day-to-day opinions about design details. ICAO typically sits here
for a national programme — once international standards are met, ICAO does not
involve itself in whether the seat pitch is 29 or 32 inches.

*Keep Informed* (low power, high interest): These stakeholders are directly affected
but lack formal authority. Passengers care enormously about the cabin experience but
cannot write the specification. Maintenance crew care about access panel locations
but rarely hold votes at critical design reviews.

*Monitor* (low power, low interest): Environmental monitoring bodies often occupy
this quadrant for programmes that comply with baseline noise and emissions standards.

## Stakeholder Landscape Across Vehicle Types

The composition and relative weight of stakeholder groups change substantially across
vehicle types. This matters because it determines which requirements dominate, which
trade-offs are acceptable, and where design effort should concentrate.

On a commercial transport, the certifying authority and the launch airline together
dominate the requirements space. The airline's operational economics drive seat count,
range, and fuel burn targets. The regulatory authority defines the certification
basis, setting structural load factors, systems reliability targets, and software
assurance levels. Passenger experience matters significantly but is filtered through
the airline's commercial judgment rather than expressed directly.

On a military strike aircraft, the structure inverts in important ways. The military
customer funds the programme and defines the operational requirements, but the using
operator — the front-line pilot, the squadron maintenance officer, the combat
logistics team — has a different priority ordering that may not be well represented
in the formal specification. In multi-service programmes like the F-35, three
separate using organisations — Air Force, Navy, Marine Corps — had different
priority orderings, and resolving the resulting tension drove a large portion of the
design compromises embedded in the programme {cite}`raymer2018`.

Small commercial UAVs operate in a compressed stakeholder space: a small set of
regulatory constraints, a clear operator, and a straightforward primary mission.
eVTOL urban air mobility sits at the opposite extreme. FAA Special Conditions
SC-VTOL, EASA SC-VTOL, local noise ordinances, air traffic management integration,
public acceptance, and vertiport infrastructure all bear on the design simultaneously
— and several of these stakeholders had not finished defining their own requirements
when the first eVTOL programmes began pursuing type certificates.

## The Customer–User Distinction

One of the most useful distinctions in any aircraft programme is between the
*customer* and the *user*. The customer is the organisation that funds the programme
and makes contractual commitments. The user is the person or organisation that
actually operates the aircraft. These are not the same, and treating them as the
same tends to produce aircraft that work on paper and fail in service.

In commercial aviation, the customer is the airline and the users are the passengers
and crew. The airline can negotiate requirements on their behalf — and launch
customers with significant leverage do this effectively — but the negotiation is
imperfect. Airlines optimise for economic efficiency; passengers optimise for comfort
and reliability; crew optimise for workability and safety margins. These objectives
sometimes point in different directions.

In government and military procurement, the gap is wider. The procuring agency — a
national ministry of defence, the Pentagon — is the customer. The using organisation
— the squadron, the depot, the operational command — is the user. Requirements
written by programme offices do not always capture the operational realities that
front-line users know from experience. The result is aircraft that meet every
specification line item on the contract but generate maintenance workarounds and
operational limitations that the programme office never anticipated.

The corrective is straightforward in principle and difficult to enforce in practice:
involve users directly in the requirements development process. This means including
front-line pilots in design reviews, not just programme managers. It means having
line maintenance technicians evaluate access panel layouts on a physical mockup, not
reviewing a CAD model on a screen. It means walking through a working day for every
primary user group and asking honestly whether the aircraft supports that job
{cite}`incose2015`.

## Voice of the Customer

The Voice of the Customer process is the structured method for capturing what
stakeholders actually need before those needs are translated into requirements. The
term comes from Quality Function Deployment (QFD), a product development framework
developed in Japan in the late 1960s and adopted widely in manufacturing and
aerospace {cite}`ullman2009`.

The process involves structured interviews with representative stakeholders,
observation of current operations — what does the existing aircraft make difficult?
— and analysis of operational data: which failures recur, and which maintenance
actions consume the most labour? The output is a set of raw, qualitative need
statements in the stakeholder's own language, not yet in engineering format.

These statements are then organised into themes using affinity diagrams and
translated into measurable objectives using QFD. The translation is where
engineering begins. "It needs to be easy to refuel" becomes "single-point
refuelling connection accessible from ground level without ground-support equipment,
maximum hose coupling torque 20 Nm, total refuelling time for full tanks less than
15 minutes." The first statement is a customer need; the second is the beginning of
a requirement.

The VoC process cannot function without a complete stakeholder map. You cannot know
whose voice to capture if you have not first identified everyone who has a voice to
give. Stakeholder analysis enables VoC, VoC enables requirements, and nothing in
the downstream engineering work — sizing, configuration selection, structural
analysis, systems integration — stands on solid ground until this chain is
established {cite}`nasa2017seh`.

```{tip}
Build your stakeholder register before your first programme review, not after.
Every major requirement churn event — the kind that drives cost growth and schedule
slippage — traces back to a stakeholder whose concerns were not identified until
late. Treat the power–interest matrix as a live document: revisit it whenever the
programme context shifts, a new regulatory requirement appears, or a new customer
comes on board.
```
