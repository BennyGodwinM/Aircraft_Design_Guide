# 2.4 — Mission Profiles & Design Drivers

An aircraft is sized by its most demanding mission. Before any weight equation is
run or any aerodynamic analysis is performed, the designer must answer one question:
what, exactly, does this aircraft need to do? The answer is the mission profile.

A mission profile is a complete description of a single representative flight from
start to finish — every phase, in sequence, with the performance conditions specified
for each phase. The sizing process then determines the minimum aircraft capable of
completing that mission while meeting all performance requirements at every phase.

This matters because different phases impose very different demands. Takeoff and
initial climb demand high thrust. Cruise demands high aerodynamic efficiency. A
sustained combat manoeuvre demands high thrust-to-weight ratio and structural
strength. Reserve fuel required after landing constrains the maximum landing weight.
The aircraft that closes on the sizing mission is the smallest one that can handle
all of these simultaneously — not the one optimised for any single phase.

## Mission Segment Definitions

The standard segments for a fixed-wing aircraft mission are {cite}`raymer2018`:

*Engine start and taxi* — ground operation at idle or low power from parking to the
runway. Fuel burn is small but non-negligible at large airports with long taxi
distances.

*Takeoff* — the accelerate–stop and rotate sequence from brake release through 35 ft
(or 50 ft, depending on the certification standard and obstacle environment). This
phase sizes the minimum field length and, in combination with approach performance,
constrains the high-lift system design.

*Climb* — from screen height to initial cruise altitude. Fuel burn during climb is
significant on long routes; on short routes, the aircraft may spend the majority of
its flight in the climb phase and never reach an efficient cruise condition.

*Cruise* — sustained level flight at the design speed and altitude. Cruise efficiency
— the product of lift-to-drag ratio and propulsive efficiency — is the dominant
driver of fuel burn on medium and long-haul routes and is the primary aerodynamic
design objective for commercial transport aircraft.

*Loiter or hold* — sustained flight at a fixed position or over a specified area,
typically at low speed and low altitude. Loiter efficiency (endurance per unit fuel)
is maximised at a different flight condition from cruise efficiency, and aircraft
designed primarily for loiter use different aerodynamic configurations from those
designed primarily for range.

*Descent and approach* — from cruise altitude to the approach fix. Descent consumes
relatively little fuel, and the altitude energy spent in the climb is dissipated
rather than recovered.

*Landing and rollout* — from the runway threshold through deceleration to stop. Field
performance at landing constrains maximum landing weight and, through the approach
speed requirement, the high-lift system performance.

*Reserve fuel* — additional fuel carried above the design mission requirement to
comply with regulatory reserve requirements or to provide a safety margin against
diversion or extended arrival sequencing. For commercial transport aircraft, ICAO
Annex rules and FAR 121 require reserves covering alternate fuel, missed approach,
and 45-minute final reserve. Reserve requirements are often the single largest driver
of total fuel load on short-haul routes — the aircraft operating a 400 km sector may
carry more reserve fuel than trip fuel {cite}`raymer2018`.

## Which Segment Sizes the Aircraft

Each aircraft type is typically dominated by one or two sizing segments. Identifying
the sizing segment before starting the sizing calculation prevents optimising for the
wrong phase.

For a commercial transport, cruise efficiency dominates total fuel burn on all but
the shortest routes. The wing and engines are sized primarily for cruise performance,
but takeoff field length determines the minimum wing area at maximum takeoff weight.
This constraint may drive the wing to be larger than the cruise-optimal size,
requiring the high-lift system to compensate for the excess area at cruise.

For a military fighter on a combat air patrol mission, the loiter phase — maintaining
a combat-ready orbit for several hours before engagement — consumes more fuel than
the transit to and from the combat zone on many mission profiles. Specific fuel
consumption at the loiter condition and the fraction of time spent loitering set the
fuel fraction as strongly as the combat radius requirement.

For a helicopter, the hover ceiling requirement — the altitude at which the aircraft
can hover in and out of ground effect at maximum gross weight in the design
hot-and-high environment — is often the segment that sizes the rotor and engine. A
rotor large enough to hover at 6,000 ft, ISA+20°C, at MTOW is also large enough to
provide efficient cruise, because hover power varies inversely with disk area and the
sizing condition is always more demanding than the cruise condition.

## Payload-Range Diagrams

The payload-range diagram is the primary tool for communicating the design trade
space to an airline customer or military operator {cite}`torenbeek2013`. It shows
the maximum payload the aircraft can carry as a function of route distance,
accounting for the fact that carrying more payload at a fixed MTOW leaves less fuel
and therefore less range.

The diagram has a characteristic shape defined by three key points:

*Point A* (maximum structural payload): the aircraft carries the maximum payload the
structure permits, filling the cabin and cargo hold. Range is limited by the fuel
capacity remaining after the payload is loaded.

*Point B* (maximum fuel and maximum payload, where achievable): some aircraft can
simultaneously carry maximum payload and maximum fuel at the cost of exceeding MTOW,
which some ferry or positioning operations permit. For many transport aircraft the
MTOW limit precludes reaching this point.

*Point C* (maximum ferry range, zero payload): the aircraft is flown with tanks full
and no payload. This is the absolute maximum range, used for delivery flights or
positioning.

Between A and C, payload is traded for range: lighter payload allows more fuel.
The slope of the A-to-C line reflects the fuel burn rate and structural weight
fractions — steeper means the design is more weight-limited.

Figure 2 shows a schematic payload-range diagram with the three characteristic
points and the structural, fuel, and MTOW constraints annotated.


````python
import matplotlib as mpl
mpl.rcParams.update({
    'figure.facecolor': 'none', 'axes.facecolor': 'none',
    'savefig.facecolor': 'none', 'savefig.transparent': True,
    'axes.edgecolor': '#888888', 'axes.labelcolor': '#cccccc',
    'text.color': '#cccccc', 'xtick.color': '#cccccc', 'ytick.color': '#cccccc',
    'grid.color': '#555555', 'font.size': 14, 'axes.titlesize': 16,
    'axes.labelsize': 14, 'xtick.labelsize': 13, 'ytick.labelsize': 13,
    'legend.fontsize': 13, 'legend.title_fontsize': 13,
    'figure.titlesize': 16, 'lines.linewidth': 2.0,
})
OKI = ['#0072B2','#D55E00','#009E73','#E69F00','#56B4E9','#CC79A7','#F0E442','#333333']
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

fig, ax = plt.subplots(figsize=(9, 6))
fig.patch.set_facecolor('white'); ax.set_facecolor('#fafafa')

# Characteristic points (range in km, payload in tonnes)
rA, pA = 5200, 19.5   # max structural payload
rB, pB = 7800, 18.0   # max fuel + structural
rC, pC = 13500, 0.0   # ferry range

r_max_struct = rA
r_full_fuel  = 3200   # range when tanks are full at MTOW with max payload

# Segments
seg1 = ([0, r_max_struct], [pA, pA])          # constant max payload
seg2 = ([r_max_struct, rB], [pA, pB])         # MTOW limit slope
seg3 = ([rB, rC], [pB, pC])                  # full-fuel slope

ax.fill_between([0]+[r_max_struct,rB,rC], [pA]+[pA,pB,pC], alpha=0.09, color=OKI[0])
ax.plot(*seg1, color=OKI[0], lw=2.5)
ax.plot(*seg2, color=OKI[0], lw=2.5)
ax.plot(*seg3, color=OKI[0], lw=2.5, ls='--')

for r, p, lbl, dy in [(rA, pA, 'A', 0.5), (rB, pB, 'B', 0.5), (rC, pC, 'C', 0.5)]:
    ax.scatter(r, p, s=100, color=OKI[1], zorder=5, edgecolors='white', lw=1.5)
    ax.text(r+120, p+dy, lbl, fontsize=11, fontweight='bold', color=OKI[1])

ax.annotate('Max structural\npayload limit', xy=(r_max_struct/2, pA),
            xytext=(r_max_struct/2, pA+1.5),
            ha='center', fontsize=9, color='#444444',
            arrowprops=dict(arrowstyle='->', color='#888888', lw=1.2))
ax.annotate('MTOW limit\n(payload traded for fuel)', xy=((rA+rB)/2, (pA+pB)/2),
            xytext=((rA+rB)/2 - 800, (pA+pB)/2 + 2.0),
            ha='center', fontsize=9, color='#444444',
            arrowprops=dict(arrowstyle='->', color='#888888', lw=1.2))
ax.annotate('Max fuel\n(zero payload ferry)', xy=(rC*0.75, pC+(pB-pC)*0.25),
            xytext=(rC*0.6, 4.5),
            ha='center', fontsize=9, color='#444444',
            arrowprops=dict(arrowstyle='->', color='#888888', lw=1.2))

ax.set_xlim(0, 15000); ax.set_ylim(0, 22)
ax.set_xlabel('Range with Design Payload (km)', fontsize=11, color='#333333')
ax.set_ylabel('Payload (tonnes)', fontsize=11, color='#333333')
ax.set_title('Payload–Range Diagram — Representative Medium-to-Long-Haul Transport',
             fontsize=12, fontweight='bold', color='#111111')
ax.tick_params(colors='#444444')
ax.grid(color='#e0e0e0', lw=0.8)
for sp in ax.spines.values(): sp.set_edgecolor('#cccccc')
plt.tight_layout(); plt.show()
````

*Figure 2 — A schematic payload–range diagram for a representative medium-to-long-
haul commercial transport. Point A is the maximum structural payload; Point B is

achievable if MTOW permits); Point C is the ferry range with empty cabin. The dashed
line from B to C shows the full-fuel constraint. Numerical values are illustrative.
Constructed from the payload-range methodology in {cite}`torenbeek2013` and
{cite}`raymer2018`.*

## Design Mission Versus the Operational Envelope

The design mission is a single set of conditions that sizes the aircraft. But
operators need their aircraft to be useful across a range of missions, not just the
design mission. Off-design performance — how the aircraft behaves when it is lighter,
shorter range, or at different altitudes than the design condition — must also be
acceptable.

An aircraft sized on a 5,000 km, 200-passenger design mission should still be
economically competitive on 1,000 km routes with 150 passengers, even if it is not
optimal at that condition. The sizing mission defines the design point; the
off-design assessment determines whether the design point was chosen sensibly for
the operational reality.

For military platforms, the design mission concept is formalised through
specification of multiple mission types, each defining a different combination of
payload, radius, and speed. The aircraft that meets all specified mission types —
simultaneously, at the same gross weight and fuel quantity — is the design that
closes. Where trade-offs are required, the relative priority of mission types is
established during the requirements process.

```{tip}
The choice of design mission is a programme decision, not an engineering decision.
Engineers can tell you what it costs in weight and fuel to extend the design range
by 10% or to add 2,000 kg of payload. They cannot tell you whether the cost is worth
paying — only the customer can answer that. Get the design mission agreed in writing
before the sizing process starts. Changing it midway is one of the most expensive
things that can happen in conceptual design {cite}`raymer2018`.
```
