# 2.2 — Needs, Goals & Objectives

A stakeholder says: "We need an aircraft that can carry a lot of cargo." That
sentence is useful. It is not a requirement. Between a raw stakeholder need and a
verifiable engineering specification lies a translation process that, when done
well, produces a clear and consistent set of technical objectives. When done badly,
it produces a requirements document full of contradictions, unmeasurable criteria,
and implicit design decisions disguised as constraints.

This section covers the translation chain — from the raw operational need to the
formal requirement ready for the traceability matrix.

## The Operational Concept

The first step is not writing requirements. It is writing the Operational Concept
document, the CONOPS {cite}`nasa2017seh`. The CONOPS describes who will use the
vehicle, in what scenarios, at what frequencies, and against what operational
environment. It is written in plain language from the user's perspective, without
reference to any specific technical solution.

A CONOPS for a short-haul regional turboprop might read: "Aircraft operates from
runways as short as 1,200 m at altitudes up to 8,000 ft density altitude. Turnaround
time between flights is 20 minutes. The aircraft is maintained at line stations with
a two-person crew and no specialist tooling. Typical route is 200–600 km, with up to
four sectors per day."

Every statement has engineering consequences. "Runways as short as 1,200 m"
constrains field performance. "8,000 ft density altitude" constrains hot-and-high
engine performance. "Two-person crew, no specialist tooling" constrains systems
accessibility and maintenance philosophy. "Four sectors per day" constrains fatigue
life targets and inspection intervals. The CONOPS is not a requirements document —
it is the environment in which requirements must make sense {cite}`ullman2009`.

A CONOPS written before any requirements allows the engineering team to identify
conflicts early. If operations says "four sectors per day" and maintenance says "each
sector requires a full external visual inspection," the conflict is visible and
resolvable before it is locked into a formal requirement. After it is locked in,
resolving it costs orders of magnitude more.

## From Need to Requirement

The translation chain has four levels. A **need** is a capability the user requires,
expressed without solution bias: "The aircraft must be able to take off from
unprepared runways." A need says what must happen, not how.

A **goal** is a more specific statement of what the design must achieve, often
including a general direction: "Field performance must support operation from
unsurfaced strips in forward-deployed environments."

An **objective** attaches a measurable target and a reference condition: "The
aircraft shall achieve a balanced field length of 900 m or less at MTOW, sea level,
ISA+15°C."

A **requirement** is an objective that has been formally reviewed, accepted, and
placed under configuration control. It can only be changed through an approved
change process. Requirements carry the weight of the contract or the regulatory
certification basis {cite}`sadraey2012`.

The hierarchy looks clean on a slide. In practice, the translation from need to
requirement is the hardest engineering work at the conceptual stage. The difficulty
is not technical — it is deciding what number to put on the objective, and why.

## Quantifying Qualitative Needs

"The aircraft must be easy to refuel" is a need. Making it a requirement means
answering: easy for whom, with what equipment, in what time? The answers come from
the CONOPS and from VoC interviews.

A structured approach is to examine the aircraft currently in service, identify the
specific pain points, and set the requirement at a level that eliminates those pain
points. If the current aircraft takes 25 minutes to refuel because the single-point
coupling is located behind a panel requiring a step ladder, and the operator's
minimum turnaround time is 20 minutes, then the requirement becomes: "Gravity and
pressure refuelling connection accessible from ground level without ground support
equipment; maximum hose coupling torque 20 Nm; full-load single-point refuelling
time ≤ 12 minutes at standard flow rate."

That requirement came from operations, not from an engineering first guess. It is
verifiable. It does not specify how the coupling is located — it specifies what it
must do and how long it must take. That is the form a good requirement takes.

## Goals, Thresholds, and Objectives in Defence Acquisition

In military and government procurement, requirements are often structured as a pair:
a **threshold** and an **objective**. The threshold is the minimum acceptable
performance — below this value, the aircraft does not meet the programme's basic
need. The objective is the desired performance — if achieved, it provides meaningful
operational benefit beyond the threshold.

This structure accommodates the trade space inherent in defence programmes. A
threshold of 600 km combat radius and an objective of 900 km tells the contractor
that 600 km is the floor, any solution exceeding 600 km is technically compliant,
and the design that achieves 900 km will be more competitive than one that achieves
601 km {cite}`sadraey2012`.

The US Department of Defense formalises this through the Joint Capabilities
Integration and Development System (JCIDS):

*Key Performance Parameters* (KPPs) are the attributes so critical that failing them
means the programme fails. They are the threshold requirements. Typical KPPs include
range, payload, survivability signature limits, and system availability.

*Key System Attributes* (KSAs) are important but not critical. Missing a KSA does
not kill the programme but will affect operational utility. Typical KSAs include
reliability targets, interoperability specifications, and training system
requirements.

*Additional Performance Attributes* (APAs) are desirable but tradeable. They may
drive contract award scoring without being contractual absolutes.

## The Danger of Over-Specifying

There is a failure mode in requirements writing almost as harmful as
under-specifying: requiring a specific solution rather than a specific capability.

"The aircraft shall have a wing swept at 35°" is a requirement that specifies a
design, not a capability. "The aircraft shall achieve Mach 0.84 cruise at FL350 at
MTOW" is a requirements statement that allows the designer to choose the wing sweep,
thickness ratio, and planform geometry that achieves the specified speed. The first
requirement artificially narrows the design space. The second leaves it open.

Over-specification is tempting when the requirements writer is also an engineer who
already has a design in mind. The antidote is to read every proposed requirement and
ask: what capability does this deliver? Can that capability be stated as a measurable
outcome without specifying how it must be achieved? If yes, rewrite it
{cite}`ullman2009`.

## Top-Level Aircraft Requirements

The first concrete output of the requirements development process is the set of
Top-Level Aircraft Requirements (TLARs). These are a small number — typically 10 to
30 — of high-level performance and operational requirements that define the mission
boundary. They are the inputs to the conceptual sizing process.

Figure 2 shows a representative TLAR set for a medium-haul commercial transport,
plotted alongside the threshold and objective values for each KPP. The gap between
threshold and objective defines the trade space available to the designer.


````python
import matplotlib as mpl
mpl.rcParams.update({
    'figure.facecolor': 'none', 'axes.facecolor': 'none',
    'savefig.facecolor': 'none', 'savefig.transparent': True,
    'axes.edgecolor': '#888888', 'axes.labelcolor': '#cccccc',
    'text.color': '#cccccc', 'xtick.color': '#cccccc', 'ytick.color': '#cccccc',
    'grid.color': '#555555', 'font.size': 14, 'axes.titlesize': 16,
    'axes.labelsize': 14, 'xtick.labelsize': 13, 'ytick.labelsize': 13,
    'legend.fontsize': 13, 'legend.title_fontsize': 13,
    'figure.titlesize': 16, 'lines.linewidth': 2.0,
})
OKI = ['#0072B2','#D55E00','#009E73','#E69F00','#56B4E9','#CC79A7','#F0E442','#333333']
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

import matplotlib.ticker as ticker

params = [
    'Design Range (km)', 'Design Payload (PAX)', 'Cruise Speed (Mach)',
    'TOFL at MTOW (m)', 'Approach Speed (kt)', 'Max Op. Altitude (ft)',
]
thresholds = [4500, 160, 0.80, 2600, 140, 41000]
objectives  = [6200, 200, 0.85, 2200, 132, 43000]

# Normalise to [0,1] for display
ranges = [max(o,t) - min(o,t) + 1e-9 for t,o in zip(thresholds, objectives)]
thr_n  = [0.0]*len(params)
obj_n  = [(o-t)/r for t,o,r in zip(thresholds,objectives,ranges)]
# If objective < threshold (e.g. TOFL, approach speed — lower is better), flip
better_lower = [False, False, False, True, True, False]
for i, bl in enumerate(better_lower):
    if bl:
        thr_n[i], obj_n[i] = obj_n[i], thr_n[i]

fig, ax = plt.subplots(figsize=(10, 5))
fig.patch.set_facecolor('white'); ax.set_facecolor('#fafafa')

y = np.arange(len(params))
bar_h = 0.35

ax.barh(y + bar_h/2, [1.0]*len(params), bar_h, color='#dddddd', label='Threshold range', zorder=2)
ax.barh(y - bar_h/2, [1.0]*len(params), bar_h, color='#bbbbbb', label='Objective range', zorder=2)

for i, (t, o, bl, rng) in enumerate(zip(thresholds, objectives, better_lower, ranges)):
    t_label = f'{t:,.0f}' if rng > 10 else f'{t:.2f}'
    o_label = f'{o:,.0f}' if rng > 10 else f'{o:.2f}'
    ax.text(0.02, i+bar_h/2, f'Threshold: {t_label}',
            va='center', fontsize=9.5, color='#333333', zorder=5)
    ax.text(0.02, i-bar_h/2, f'Objective: {o_label}',
            va='center', fontsize=9.5, color=OKI[0], zorder=5)

ax.set_yticks(y); ax.set_yticklabels(params, fontsize=10.5, color='#222222')
ax.set_xticks([]); ax.set_xlim(0, 1.0)
ax.set_title('Representative TLAR Set — Medium-Haul Commercial Transport\n'
             'Threshold (minimum acceptable) vs. Objective (operationally desired)',
             fontsize=12, fontweight='bold', color='#111111')
for sp in ['top','right','bottom']: ax.spines[sp].set_visible(False)
ax.spines['left'].set_edgecolor('#cccccc')
plt.tight_layout(); plt.show()
````

*Figure 2 — Representative TLARs for a medium-haul commercial transport, showing
threshold and objective values for each Key Performance Parameter. For parameters
where lower is better (field length, approach speed), the threshold is the least
stringent acceptable value and the objective is the more demanding target.
Numerical values are illustrative; real TLARs are programme-specific and
commercially sensitive. Constructed from methodology in {cite}`raymer2018` and
{cite}`sadraey2012`.*

Typical TLARs for a commercial transport include design range (km or nm), design
payload (kg or passengers), cruise speed (Mach number), maximum operating altitude
(ft), approach speed (knots), balanced field length at MTOW (m), and maximum
landing weight as a fraction of MTOW. Each TLAR has a single numeric value with
clearly defined reference conditions — ISA, sea level, or specified altitude.

TLARs are not the complete design requirements set. They are the requirements that
size the aircraft — that determine the gross weight, the wing area, the engine
thrust, and the fuel fraction. Every other requirement is then checked for
compatibility against the sized configuration and fed into the preliminary and
detailed design phases.

```{tip}
Write your TLARs before running any sizing code. Every sizing equation in Raymer
or Roskam takes TLARs as inputs. Sizing a design before the TLARs are agreed is
equivalent to solving an equation before deciding what the equation says. It
produces a number, but the number does not mean anything {cite}`raymer2018`.
```

```{note}
The threshold/objective structure used in US defence procurement (KPP, KSA, APA)
has equivalents in other acquisition systems. The UK MOD uses a similar structure in
its CADMID (Concept, Assessment, Demonstration, Manufacture, In-service, Disposal)
framework. NATO programmes align with STANAG 4685 for capability requirements
documentation.
```
