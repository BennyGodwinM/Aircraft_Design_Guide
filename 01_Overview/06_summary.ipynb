# 1.6 — Summary

Aircraft design is a structured process of creating a flying machine that fulfils an identified need under real-world constraints. It is iterative, interdisciplinary, and bounded by competing objectives. There is never one correct answer — only better and worse trade-off decisions.

| Principle | Implication |
|-----------|-------------|
| Iterative process | Expect to revisit and revise — this is the method |
| No single right answer | Focus on defensible trade-offs, not perfection |
| Educated judgment | Document assumptions so they can be challenged |
| Interdisciplinary | Decisions ripple — communicate across disciplines early |
| Real hurdles | Weight, requirements conflict, certification, cost, and uncertainty are structural features of the field |

Each subsequent chapter will peel back one layer of this complexity, building toward a complete picture of how real aircraft are conceived, analysed, sized, and refined.
