# 1.7 A History of Aircraft Design

Aircraft design did not spring into existence fully formed. It evolved over more than a century through an iterative process driven by war, commerce, competition, and — above all — the enduring human desire to fly. Understanding that history is more than an exercise in nostalgia: every major advance in design methodology was a response to a real engineering failure, a military requirement, or an economic constraint. The same pressures that shaped the Wright Flyer of 1903 are, in transformed guise, still shaping the airliners and uncrewed systems of today.

This notebook traces the development of aircraft design from its theoretical origins in the early nineteenth century to the digital, data-driven practice of the twenty-first. It also examines how geography, politics, and industrial culture have produced distinct national and regional design traditions that continue to colour the aerospace industry.

## 1. The Theoretical Pioneers (Before 1900)

### Sir George Cayley and the Science of Flight

The intellectual history of aircraft design begins not at Kitty Hawk but in the English countryside of the early nineteenth century. **Sir George Cayley** (1773–1857) is widely regarded as the father of aeronautics. Working decades before the petrol engine existed, Cayley identified the four fundamental forces of flight — lift, drag, thrust, and weight — and was the first to appreciate that they must be handled by separate surfaces. His 1804 model glider, with a fixed main wing, a moveable tail for pitch and yaw, and a paddle wheel for thrust, established the configuration that virtually every fixed-wing aircraft still follows.

Cayley's design method was empirical and analytical in equal measure. He built small models, measured their behaviour, and reasoned from first principles — a process that anticipates the modern design spiral. His 1853 full-scale glider, which carried his coachman across a Yorkshire valley, was the first piloted, heavier-than-air flight of a fixed-wing vehicle.



### Lilienthal and the Art of Gliding

**Otto Lilienthal** (1848–1896) advanced Cayley's ideas from the drawing board to the air. A German engineer, Lilienthal made more than two thousand gliding flights between 1891 and his fatal crash in 1896, systematically measuring lift and drag on cambered aerofoils and publishing his results in *Birdflight as the Basis of Aviation* (1889). His data tables — the first systematic aerodynamic database — were used directly by the Wright Brothers a decade later.

Lilienthal's method was body-shifting control: he altered the aircraft's attitude by swinging his legs and torso to shift the centre of gravity. It worked for pitch and roll in calm air, but it was inadequate for adverse conditions — a lesson the Wrights drew sharply from his death.



### Other Key Figures of the Pre-Wright Era

**Octave Chanute** (1832–1910), a French-born American civil engineer, was the great synthesiser of nineteenth-century aeronautical knowledge. His 1894 book *Progress in Flying Machines* surveyed every serious attempt to build a flying machine and drew out what each had demonstrated. Chanute corresponded extensively with the Wrights and visited their camp at Kitty Hawk.

**Samuel Langley** (1834–1906), Secretary of the Smithsonian Institution, built steam-powered unmanned models that flew successfully in 1896 and received US government funding to build a full-scale machine. His *Aerodrome* failed on its launch attempts in October and December 1903 — just nine days before Kitty Hawk — partly because of a deficient launching catapult and partly because the design had no effective means of maintaining longitudinal stability in gusty conditions. The contrast with the Wrights' methodical approach to control is instructive.

## 2. The Wright Brothers and the Birth of Controlled Flight (1903–1914)

### Method Over Intuition

Orville and Wilbur Wright succeeded where better-funded contemporaries failed because they were, before anything else, rigorous engineers. They began not by building aircraft but by reading — especially Lilienthal's data — and then testing. When they found Lilienthal's lift coefficients inconsistent, they built a wind tunnel in their Dayton bicycle shop in 1901 and spent the winter systematically testing over two hundred aerofoil and wing planform shapes. Their data was more accurate than anything previously published, and it led them to aerofoils with higher camber and aspect ratios than their predecessors had used.

Crucially, the Wrights identified that the central unsolved problem of flight was **control**, not lift. Langley had lift; what killed his Aerodrome was the absence of effective three-axis control. The Wrights devised **wing warping** — twisting the outer panels of the biplane wing in opposite senses — to generate differential lift for roll control. Combined with a moveable forward elevator (pitch) and a single rear rudder linked to the wing-warp mechanism (yaw), this gave them the first complete three-axis control system.



### From Flyer to Practical Aeroplane

The 1903 Flyer was unstable and barely controllable by any pilot other than the brothers. The Wright A of 1908, demonstrated in France by Wilbur, was a revelation to European observers — smooth, banking turns and sustained circuits lasting minutes, not seconds. Yet European designers quickly overtook the Wrights by abandoning the canard (forward elevator) layout and adopting the conventional tail-aft configuration, which proved more stable and easier to fly.

**Louis Bleriot's** monoplane crossing of the English Channel in July 1909 marked the decisive public demonstration that the aeroplane was a practical vehicle, not a laboratory curiosity. Bleriot's tractor monoplane — engine and propeller at the front, pilot in the middle, tail at the rear — became the template that almost every subsequent single-engine aircraft would follow.

## 3. World War I — The First Aviation Revolution (1914–1918)

### War as an Accelerant

No force has accelerated aircraft development more rapidly than war. In August 1914 the typical military aeroplane was a fragile, unarmed two-seat reconnaissance machine of wood, fabric, and piano wire, capable of perhaps 120 km/h and barely able to climb above 2 000 m. By November 1918 fighters were capable of 220 km/h at 6 000 m, with synchronised machine guns, oxygen systems, and radio equipment. Four years of total war compressed what might have been two decades of peacetime development.

### Key Technical Advances

**The synchronised gun** (1915) transformed air combat and forced designers to prioritise forward visibility and manoeuvrability. The Fokker Eindecker, fitted with Anthony Fokker's mechanical interrupter gear, dominated the Western Front for months — the so-called *Fokker Scourge* — until the Allies developed their own synchronisation systems.



**The rotary engine** gave early fighters a high power-to-weight ratio but introduced severe gyroscopic effects that influenced handling — the Sopwith Camel's notoriously vicious left-turn tendency was a direct consequence. The eventual move to water-cooled in-line engines (the Hispano-Suiza V-8, the Rolls-Royce Eagle) pointed the way toward the high-powered, high-altitude fighters of the 1930s.

**Structural refinement** proceeded rapidly. The wire-braced biplane structure reached a high point of development. Hugo Junkers' all-metal, internally-braced monoplane designs — particularly the Junkers J 1 of 1915 — anticipated the stressed-skin construction that would dominate the following two decades, though they were too heavy for fighters at the time.

### Design Process in Wartime

WWI design was overwhelmingly empirical. Test pilots evaluated prototypes and modifications were made by rule of thumb rather than analysis. There was no wind-tunnel testing of full configurations; structural loads were estimated by simple beam theory and corrected by experience. The design cycle — from drawing to first flight — could be as short as a few weeks. Speed of iteration mattered more than analytical depth.

## 4. The Interwar Golden Age (1919–1939)

### Streamlining and the Monoplane Revolution

The two decades between the wars produced the most dramatic change in the physical form of the aircraft that has ever occurred in a comparable period. The biplane of 1919 gave way to the all-metal stressed-skin monoplane of 1939 — a transformation driven by aerodynamic theory, new materials, more powerful engines, and commercial competition.

The **NACA cowling** (1927) reduced the aerodynamic drag of radial engines by as much as 60 percent in some installations — a single component that transformed the economics of air transport. NACA's systematic programme of low-drag aerofoil development, beginning in the late 1920s and culminating in the laminar-flow four-digit and five-digit series of the 1930s, gave designers a reliable, data-backed aerofoil library for the first time.



### The Birth of the Modern Airliner

The **Boeing 247** (1933) and the **Douglas DC-3** (1935) define the modern airliner template. The 247 was the first aircraft to combine all-metal stressed-skin construction, retractable undercarriage, variable-pitch propellers, de-icing, and full controllability on one engine into a single, commercially viable package. United Air Lines ordered sixty, effectively locking out competitors and prompting TWA to commission Douglas to design something better. The result was the DC-1/DC-2 series, refined into the DC-3 — perhaps the most economically successful aircraft ever built.

The design philosophy behind both aircraft was decisively American: standardised components, ease of maintenance, and economic viability. European contemporaries — the Junkers Ju 52 in Germany, the Armstrong Whitworth Atalanta in Britain — reflected different priorities: the Ju 52's corrugated Duralumin skin was structurally conservative but drag-inducing; the Atalanta prioritised range on imperial routes over passenger comfort and operating cost.

### The Air Racer as a Design Driver

The Schneider Trophy seaplane races (1913–1931) and the National Air Races in the United States drove aerodynamic refinement in ways that no commercial programme could justify. The Supermarine S.6B, which won the Schneider Trophy outright in 1931, reached 655 km/h and established the aerodynamic knowledge base — wing loading, cooling, aerofoil profiles at high speed — that Reginald Mitchell drew on directly in designing the Spitfire.

## 5. World War II — The Great Leap Forward (1939–1945)

### Six Years, Sixty Years of Progress

The Second World War compressed aeronautical progress more violently than even WWI had done. Aircraft that were state of the art in 1939 were obsolete by 1942; designs that seemed futuristic in 1942 were in squadron service by 1944. By 1945 the piston engine had been pushed to its ultimate development, jet propulsion had entered operational service, and the engineering methods that would define the postwar decades had been established.

### Piston Engine Supremacy — and Its Limits

The great piston-engined fighters of WWII — the Spitfire, the Bf 109, the P-51 Mustang, the Zero, the F4U Corsair — represented the aerodynamic art taken close to its absolute limits within the constraint of a reciprocating engine. The P-51's laminar-flow wing, designed using the NACA 45-100 series, reduced profile drag by 30 percent compared with conventional aerofoils at cruise speed; combined with an efficient radiator duct arrangement, it gave the Mustang the range to escort bombers deep into Germany.



### The Jet Age Begins

Frank Whittle filed his jet engine patent in 1930; Hans von Ohain independently arrived at the same concept in Germany in the mid-1930s. The first jet-powered flight was made by the Heinkel He 178 on 27 August 1939 — three days before Germany invaded Poland. The Messerschmitt Me 262, the world's first operational jet fighter, entered Luftwaffe service in limited numbers in 1944. Its 850 km/h top speed was some 200 km/h faster than any Allied piston fighter, yet it came too late and in insufficient numbers to affect the outcome of the war.



### Advances in Design Method

WWII saw the first systematic use of **structural analysis** beyond simple beam theory. Finite-element methods did not yet exist, but the stress offices of de Havilland, Supermarine, Focke-Wulf, and Boeing employed large teams of human "computers" solving systems of equations by hand. Wind-tunnel programmes expanded enormously. The NACA at Langley Field ran continuous programmes testing configuration after configuration for the Army Air Forces and Navy.

The German research establishment — the DVL, AVA Gottingen, and research groups at the aircraft companies — made particularly important advances in high-speed aerodynamics. Work by Busemann, Betz, and others on swept wings to delay compressibility effects was captured by Allied technical teams after the war and directly influenced the design of the F-86 Sabre and the Boeing 707.

## 6. The Jet Age and the Cold War (1945–1975)

### Breaking the Sound Barrier

On 14 October 1947, Captain Charles "Chuck" Yeager flew the Bell X-1 to Mach 1.06 at 13 000 m over the Mojave Desert. The sound barrier, which had claimed several lives and seemed to many to be a genuine physical limit, was broken. Within a decade, supersonic flight was routine for military aircraft and the design problems that had seemed intractable — shock-wave drag, control reversal near Mach 1, inertial coupling — had been solved.

The **area rule**, discovered by Richard Whitcomb at NACA in 1952, resolved the wave drag crisis that had stalled the early supersonic fighter programme. By constricting the fuselage cross-section where the wings added cross-sectional area — producing the characteristic "Coke bottle" shape — wave drag could be reduced by as much as 60 percent. The Convair F-102 Delta Dagger, originally designed without the area rule, failed to reach Mach 1 in level flight; redesigned with a waisted fuselage, it exceeded Mach 1 comfortably.

### Commercial Aviation Transforms

The **de Havilland Comet** (1952) was the world's first commercial jet airliner. It halved Atlantic crossing times and demonstrated the commercial viability of jet travel. Its catastrophic pressure-cabin fatigue failures in 1954 — caused by crack propagation from the square window corners — became the defining case study in structural engineering education and transformed the industry's approach to damage-tolerant design.

The **Boeing 707** (1958) established the formula for the modern jetliner: four podded engines on swept wings, T-tail or conventional tail, pressurised fuselage of large diameter. Boeing's gamble in building the 367-80 prototype at corporate expense — a decision that staked the company's survival — paid off spectacularly. The design philosophy was unabashedly commercial: low risk, high reliability, and margin for growth through derivatives.



### The Wide-Body Era and Concorde

The **Boeing 747** (1969) was an act of extraordinary industrial ambition. At the time of its design, no high-bypass turbofan engine existed that could power it — the Pratt & Whitney JT9D had to be developed in parallel. The 747's double-bubble fuselage cross-section, twin aisles, and upper deck were unprecedented in their scale. Its design process required the invention of new CAD and manufacturing methods: the Everett factory built to house it was the largest building by volume in the world.

**Concorde** (first flight 1969, service entry 1976) represented a different philosophy: performance at any cost. The British-French programme spent approximately £1.3 billion (in 1976 money) to produce fourteen operational aircraft. The ogival delta wing, variable-geometry engine intakes, and drooping nose visor were all elegant engineering solutions to genuine supersonic cruise problems. That it was economically unviable from the outset was a design failure, not an engineering one — the mission specification, driven by national prestige rather than market analysis, was untenable.

## 7. The Computational Era (1970s–1990s)

### CAD, CFD, and FEA Transform the Discipline

The digital computer did not merely speed up existing design methods: it made entirely new methods possible. Three computational tools reshaped aircraft design fundamentally in the final quarter of the twentieth century.

**Computer-Aided Design (CAD)** replaced drawing boards from the mid-1970s. Early CAD systems were wireframe and 2D; solid modellers and surface modellers capable of describing complex aerodynamic shapes emerged in the 1980s. The **Boeing 777** (1994) was designed entirely in CATIA — the first major commercial aircraft to go from contract to rollout without a single physical mock-up of the full aircraft. This eliminated weeks of mock-up-related rework and allowed manufacturing tolerances to be tightened dramatically.

**Computational Fluid Dynamics (CFD)** began with simple panel methods (1960s) and progressed through Euler-equation solvers (1980s) to Reynolds-Averaged Navier-Stokes (RANS) codes (1990s). CFD did not eliminate wind-tunnel testing but it radically changed what the tunnel is used for: instead of exploring the design space, the tunnel now validates specific configurations that CFD has already screened. This reduced the number of wind-tunnel entries per programme from hundreds to tens.

**Finite Element Analysis (FEA)** transformed structural design. Before FEA, complex structures were analysed by decomposing them into idealised elements (beams, panels, shear webs) and applying semi-empirical methods. FEA allowed the entire airframe to be represented as a mesh of elements with stiffness matrices assembled and solved simultaneously, giving far more accurate prediction of stress distribution and identifying failure modes that simplified analyses would miss.

### The Stealth Revolution

The **Lockheed F-117 Nighthawk** (first flight 1981, declassified 1988) demonstrated that a design constraint never previously considered — radar cross-section — could be incorporated into the design process as a primary driver. The faceted surface geometry of the F-117 was a consequence of the limitations of early radar signature prediction codes, which could only handle flat surfaces. The **Northrop B-2 Spirit** (1989) used later smooth-surface codes to achieve lower radar signature with a more aerodynamically efficient blended wing-body form.

### Fly-by-Wire and Active Control

**Fly-by-wire (FBW)** flight control, in which mechanical linkages between cockpit controls and control surfaces are replaced by digital computers and electrical signalling, became standard in both military aircraft (F-16, 1978) and commercial transports (Airbus A320, 1988). FBW enabled artificial stability augmentation — aircraft like the F-16 and the Eurofighter Typhoon are aerodynamically unstable and can only be flown with computer assistance — allowing designers to relax stability constraints in favour of agility or aerodynamic efficiency.

## 8. The Composite and Data-Driven Era (2000–Present)

### Carbon Fibre Transforms the Airframe

The **Boeing 787 Dreamliner** (first flight 2009) was the first commercial airliner with a primary structure predominantly of carbon-fibre-reinforced polymer (CFRP): approximately 50 percent by weight. The Airbus A350 XWB (2013) followed at 53 percent composite. The shift to composites brought a 20–25 percent reduction in structural weight compared with equivalent aluminium structures, but it required a complete rethinking of the manufacturing and maintenance ecosystem: autoclave curing of enormous fuselage sections, new NDT techniques, new repair procedures.



### Model-Based and Digital Engineering

Contemporary aircraft design is increasingly practised through **Model-Based Systems Engineering (MBSE)**, in which a single authoritative digital model of the aircraft — its geometry, systems, loads, mass properties, and performance — is maintained throughout the programme from requirements capture to in-service support. The **digital twin** concept extends this: a continuously updated digital replica of each individual aircraft, fed by in-service sensor data, enables condition-based maintenance and design feedback loops that were previously impossible.

**Multidisciplinary Design Optimisation (MDO)** routinely couples aerodynamic, structural, and propulsion analyses in automated optimisation loops. A wing redesign that once required months of sequential analysis by different specialist teams can now be explored in hours in a parametric MDO framework.

### eVTOL and Urban Air Mobility

The most recent design revolution is the emergence of electric and hybrid-electric vertical take-off and landing (eVTOL) aircraft for urban air mobility. Enabled by advances in battery energy density, power electronics, and distributed electric propulsion, dozens of eVTOL designs — ranging from large multirotor platforms to winged tilting-propeller configurations — entered development during the 2010s. Their design challenges are distinct from those of conventional aircraft: battery mass fraction dominates the weight budget, hover efficiency trades directly against cruise efficiency, and acoustic signature is a certification and social acceptance constraint with no legacy regulatory framework.

## 9. Regional and National Design Traditions

Aircraft design does not occur in a cultural vacuum. National industrial traditions, military procurement cultures, geographic constraints, and institutional legacies have produced distinct approaches that remain visible today.

### The United States: Pragmatism and Industrial Scale

American aircraft design has been characterised, above all, by its commercial orientation. Boeing, Douglas, Lockheed, Grumman, and their successors designed primarily to meet quantifiable performance and economic requirements: payload-range, block fuel per seat, direct operating cost. This produced aircraft of great reliability and producibility — the DC-3, the B-17, the Boeing 747 — that were not always the most aerodynamically refined but were extraordinarily successful in the market. NACA and later NASA provided a uniquely capable public research infrastructure: the low-drag aerofoil families, the area rule, and supercritical wing sections all originated from government-funded research that was freely shared with industry.

### Britain: Elegance and Incremental Refinement

British aircraft design has historically emphasised aerodynamic refinement, structural elegance, and the mastery of difficult problems — sometimes at the expense of producibility and economics. The Spitfire's elliptical wing was aerodynamically near-optimal but expensive and slow to manufacture. The de Havilland Comet and the Concorde were daring technical achievements that preceded the market they sought to create. British engine design — Rolls-Royce in particular — has consistently led in aero engine technology and remains a world leader.

The Royal Aircraft Establishment (RAE) at Farnborough was the institutional locus of British aeronautical research from 1917 onwards, producing fundamental contributions to boundary-layer theory, transonic aerodynamics, and aeroelasticity. Its absorption into the Defence Evaluation and Research Agency (DERA) and QinetiQ from the 1990s reflected the shift toward a more commercial research model.

### Germany: Analytical Rigour and Structural Innovation

German aerospace engineering has been defined by its analytical depth and structural originality. Hugo Junkers introduced all-metal construction and internal bracing in the 1910s; Claudius Dornier pioneered the flying boat; Ernst Heinkel and Willy Messerschmitt pushed aerodynamic refinement to competitive extremes in the 1930s and 1940s. The Gottingen aerodynamic research group — particularly the work of Ludwig Prandtl and his students on boundary layers, induced drag, and wing theory — provided the theoretical foundation for modern aerodynamics.

Post-WWII, Germany's aerospace capacity was suppressed for a decade and then rebuilt within the NATO framework. Messerschmitt-Bolkow-Blohm (MBB) became one of the founding partners of Airbus, contributing particularly to structures and systems engineering. The German contribution to Airbus — careful, thorough, and analytically intensive — has been essential to the programme's success.

### France: State Direction and Technical Ambition

French aviation has been shaped by a tradition of state direction and public–private partnership unusual even by European standards. The nationalisations of the 1930s created SNCASE, SNCASO, and other state enterprises; the postwar period produced Sud-Aviation (later Aerospatiale, now Airbus Group). Dassault Aviation, though nominally private, has operated in intimate relationship with the French state through military contracts since Marcel Bloch (Marcel Dassault) designed fighter aircraft in the 1930s.

French design has been marked by a willingness to accept technical risk for performance gain — Concorde being the most extreme example, but the delta wing Mirage series and the Airbus A320's digital fly-by-wire system (the most advanced commercial FBW of its time) show the same tendency. The Centre d'Essais en Vol (CEV) and ONERA have provided strong research infrastructure.

### The Soviet Union and Russia: Quantity, Robustness, and Parallel Development

Soviet aircraft design operated within a system that emphasised parallel development — multiple design bureaux (OKB) competed simultaneously on each requirement, with the state selecting the winner for production. This produced a very different dynamic from Western practice: each OKB — Tupolev, Ilyushin, Mikoyan-Gurevich (MiG), Sukhoi, Yakovlev, Antonov — developed a distinct house style, and the competitive pressure between them drove rapid development.

Soviet design philosophy prioritised robustness over refinement, high-volume manufacturability over finish, and operational simplicity over performance ceiling. The MiG-15 and MiG-21 were produced in the tens of thousands; the An-124 Ruslan and An-225 Mriya remain the largest cargo aircraft ever built. Russian aviation continues in this tradition, with the Sukhoi Superjet and Irkut MC-21 the primary current civil programmes.

### Japan: Precision Manufacturing and Regulated Innovation

Japanese aviation was effectively suppressed from 1945 to the mid-1950s under the Allied occupation. When it resumed, the industry rebuilt within a framework of technology licensing and subcontracting for American primes — Mitsubishi, Kawasaki, and Fuji all became major subcontractors for Boeing and Lockheed. The **Mitsubishi SpaceJet** (formerly MRJ) regional jet programme, launched in the 2000s, has been the most ambitious Japanese attempt to develop a fully indigenous commercial transport, though it has faced substantial certification delays.

Japan's manufacturing precision — inherited from its automotive and electronics industries — has made Japanese aerospace suppliers world leaders in composite fuselage panels and wing components. Approximately 35 percent of the Boeing 787 airframe by weight is manufactured in Japan.

### Italy: Craftsmanship and Seaplane Heritage

Italian aviation in the interwar period was defined by the seaplane racing tradition and by designers such as Mario Castoldi (whose Macchi M.C.72 set an unbroken propeller-driven speed record of 709 km/h in 1934) and Gianni Caproni (a pioneer of large multi-engine bombers). Postwar Italian aerospace rebuilt around Alenia Aeronautica (now Leonardo Aero) and Piaggio, with a particular strength in turboprop regional aircraft and trainer jets (the Aermacchi MB-326 and MB-339).

Leonardo is a full partner in the Eurofighter Typhoon and ATR programmes, and manufactures fuselage sections for the Boeing 787. Italian design retains a characteristic attention to aesthetic finish alongside aerodynamic performance — a quality visible in the Piaggio P.180 Avanti's radical pusher-propeller, canard, and laminar-flow wing combination.

## 10. Recurring Themes and Lessons

Looking across the full arc of aircraft design history, several themes recur with striking consistency.

**War accelerates; commerce sustains.** Every major step-change in performance has been driven by military urgency. But the technologies developed under military pressure — the jet engine, the gas turbine, swept wings, fly-by-wire, composites — have invariably found their most widespread application in commercial aviation, where sustained operation over millions of cycles demands a reliability and economy that military programmes rarely require.

**The winning design is rarely the most technically advanced.** The DC-3 was not the most aerodynamically sophisticated aircraft of 1935; the Boeing 747 was not the boldest concept under consideration in the late 1960s. What made these aircraft successful was the quality of the requirement definition, the judgement to incorporate proven technology rather than untested innovation, and the robustness of the design to derivative development. The lesson is one that Raymer states explicitly: the design process begins with the mission, not the technology.

**Failures teach more than successes.** The Comet's pressure-cabin failures established fail-safe structural design philosophy. The DC-10 cargo door accidents produced the industry's modern approach to catastrophic-failure prevention. The Lockheed L-188 Electra's propeller-whirl-mode oscillation led to a complete re-evaluation of engine mount stiffness requirements. Each failure, tragic in its immediate consequences, built the regulatory and engineering framework that makes modern commercial aviation the safest mode of transport ever devised.

**Computational tools amplify human judgement — they do not replace it.** CFD can predict pressure distributions to four significant figures, but it cannot tell the engineer which configuration to analyse. MDO can find the minimum of an objective function, but it cannot define that function. The history of aircraft design is, at its core, a history of human judgement applied to ill-defined problems under conditions of uncertainty — a condition that no increase in computational power has yet fundamentally altered.
