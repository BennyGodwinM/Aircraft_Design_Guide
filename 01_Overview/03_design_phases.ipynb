# 1.3 — The Three-Phase Design Process

Aircraft design is organised into three progressive phases, each narrowing the solution space while increasing the fidelity of analysis {cite}`raymer2018` {cite}`torenbeek2013`.

```{dropdown} Phase 1 — Conceptual Design
The broadest phase. The designer asks: *What should it look like? What size? What propulsion type?*

**Tools:** historical data, statistical weight methods, back-of-envelope sizing.

**Output:** a feasible, sized concept with estimated performance.

**Freedom to change:** very high — nothing has been built, and changes cost only engineering time.
```

```{dropdown} Phase 2 — Preliminary Design
The configuration is essentially frozen; now it is refined and validated. The designer asks: *Does it really work? Can it be built? Do the subsystems integrate?*

**Tools:** computational analysis, wind tunnel testing, detailed weight estimation.

**Output:** a design ready for manufacturing planning.

**Freedom to change:** moderate — decisions made here are difficult and costly to reverse.
```

```{dropdown} Phase 3 — Detailed Design
Every component is fully defined for manufacturing. The designer asks: *How exactly is each part made? How do they assemble?*

**Tools:** full 3D CAD, FEA, fatigue analysis, qualification testing.

**Output:** production drawings and specifications.

**Freedom to change:** very low — tooling may already exist, and changes are extremely expensive.
```

```{warning}
The cost of changing a design decision grows exponentially as the design matures. A configuration change at the conceptual phase costs days of engineering time. The same change at the detailed design phase may cost millions of dollars and months of schedule {cite}`raymer2018`. This is why conceptual design, despite its apparent imprecision, is the most strategically important phase.
```

## Cost of Change Across Phases

````python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['savefig.facecolor'] = 'none'

fig, ax = plt.subplots(figsize=(8.5, 4.5))
fig.patch.set_facecolor('none')
ax.set_facecolor('none')

phases = ['Conceptual\nDesign', 'Preliminary\nDesign', 'Detailed\nDesign', 'Production']
costs  = [1, 10, 100, 1000]
x = np.array([0, 1, 2, 3])

for xi in [0.5, 1.5, 2.5]:
    ax.axvline(xi, color='#99aabb', lw=0.8, linestyle='--', alpha=0.4, zorder=0)

ax.semilogy(x, costs, '-', color='#1565c0', lw=2.8, zorder=3, solid_capstyle='round')
ax.semilogy(x, costs, 'o', markersize=13,
            markerfacecolor='none', markeredgecolor='#1565c0',
            markeredgewidth=2.5, zorder=4, linestyle='none')

mult_labels = ['×1', '×10', '×100', '×1,000']
for xi, yi, ml in zip(x, costs, mult_labels):
    ax.annotate(ml,
        xy=(xi, yi), xytext=(xi, yi * 4.0),
        fontsize=10, ha='center', color='#c0392b', fontweight='bold',
        arrowprops=dict(arrowstyle='->', color='#c0392b', lw=1.2))

for i, ph in enumerate(phases):
    ax.text(i, 0.15, ph, fontsize=9.5, ha='center', va='center',
            color='#1565c0', multialignment='center', fontweight='semibold')

ax.set_xlim(-0.55, 3.55)
ax.set_ylim(0.10, 9000)
ax.set_xticks([])
ax.set_ylabel('Relative cost of a change (log scale)', fontsize=10, color='#334455')
ax.yaxis.grid(True, which='major', linestyle='-', alpha=0.12, color='#667788')
ax.spines[['top', 'right', 'bottom']].set_visible(False)
ax.spines['left'].set_color('#99aabb')
ax.tick_params(axis='y', colors='#334455', labelsize=9)

plt.tight_layout()
plt.show()
````

*Figure 2 — Relative cost of incorporating a design change grows by roughly an order of magnitude with each successive phase. Source: {cite}`raymer2018`, Ch. 2.*
