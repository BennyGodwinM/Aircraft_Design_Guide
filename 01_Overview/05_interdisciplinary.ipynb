# 1.5 — The Interdisciplinary Nature of Aircraft Design

Aircraft design cannot be decomposed into a set of independent sub-problems and then reassembled.
Every major engineering discipline is tightly coupled to every other {cite}`raymer2018` {cite}`torenbeek2013`.
A change in aerodynamics propagates to structures, which propagates to weight, which changes the required thrust, which changes fuel consumption, which changes range — restarting the entire sizing loop.

This coupling is not an inconvenience to be managed away; it is the defining characteristic of aircraft design.
A wing that is aerodynamically optimal is rarely structurally efficient.
A structurally efficient wing is rarely compatible with the maintenance access requirements.
The discipline that "wins" in any specific trade-off is determined by the mission, the market, and the design constraints — not by any single engineer.

In industry, multidisciplinary teams work in parallel with frequent integration reviews to keep disciplines synchronised.
No single engineer solves an aircraft.
The aircraft emerges from thousands of disciplinary decisions, made in negotiation, over years.

## The core disciplines

The following sections describe each major discipline and its principal coupling to the others.


### Aerodynamics

Aerodynamics determines the lift, drag, and moment characteristics of the external configuration.
It is the discipline that sets the "design envelope" — the bounds of flight speed, altitude, and angle of attack within which the aircraft must operate safely.

**Principal outputs to other disciplines:**
- Lift distribution → structural loads → wing spar sizing
- Drag polar ($C_L$ vs $C_D$) → performance (range, climb, ceiling) via Breguet
- Pitching moment $C_m$ → horizontal tail sizing and trim drag
- Control surface aerodynamics → stability and control derivatives

**Principal inputs from other disciplines:**
- Wing geometry (span, chord, sweep, twist, taper) from structures/layout
- Nacelle position and shape from propulsion integration
- Landing gear fairing shape from systems layout

**Key aerodynamic parameters:**
- Lift coefficient $C_L = L / (q S)$
- Parasite drag coefficient $C_{D_0}$ (typically 0.015–0.025 for clean jets)
- Induced drag factor $k = 1/(\pi e AR)$ where $e$ is Oswald efficiency (typically 0.75–0.90)
- Maximum $L/D$ at cruise (15–22 for modern transports)

```{note}
Modern transports cruise at $C_L \approx 0.5$–$0.6$ and $M \approx 0.78$–$0.86$, near the onset of wave drag.
Transonic aerodynamic design — managing the shock system that forms on the wing upper surface — is one of the most computationally intensive aspects of aircraft design and has improved substantially with the advent of CFD and multi-point optimisation {cite}`drela2014`.
```


### Structural design

The structure must carry aerodynamic and inertial loads through the entire flight envelope — including gusts, manoeuvres, and ground handling — with adequate fatigue life and damage tolerance, at minimum weight.

**Principal loads:**
- Bending moment and shear at the wing root (determined by lift distribution and fuel/payload weight)
- Fuselage bending (from tail loads and payload distribution)
- Pressurisation loads (hoop stress in the fuselage skin: $\sigma = pR/t$)
- Landing impact loads (typically 2.0–2.5 g design load factor at MLW)
- Torsion (from aerodynamic pitching moment and control surface deflection)

**Materials hierarchy (roughly by application):**
- Wing box: carbon fibre reinforced polymer (CFRP) or aluminium alloy (7075-T6)
- Fuselage frames and skin: CFRP (787, A350) or aluminium alloy (737, A320)
- High-temperature zones (nacelle, APU bay): titanium or nickel superalloys
- Secondary structure, fairings: fibre-glass composite, thermoplastics

**Coupling to weight:**
Structural weight is the single largest component of empty weight — roughly 22–27% of MTOW for a modern transport.
A 1% reduction in structural weight fraction reduces fuel burn by roughly 0.5% on a typical mission, compounding over the entire fleet life into billions of dollars of savings {cite}`torenbeek2013`.

**Fatigue and damage tolerance:**
FAR/CS-25 requires that transport aircraft be designed to safe-life or damage-tolerant criteria.
The Boeing 737 famously suffered the 1988 Aloha Airlines accident (fuselage upper section separation) due to fatigue crack coalescence at lap-joint fastener holes — a failure that reshaped how multiple-site damage is analysed and inspected.


### Propulsion

The propulsion system converts fuel chemical energy into thrust (or shaft power) and determines fuel consumption through the specific fuel consumption (SFC or TSFC).

**Engine–aircraft coupling points:**
- Thrust lapse with altitude and speed sets the T/W budget on the constraint diagram
- Engine weight and nacelle drag enter directly into the weight and aerodynamic equations
- Nacelle position affects wing aerodynamics (interference drag, jet-flap effect), structural loads (bending relief or penalty), and ground clearance
- Bleed air extraction for environmental control system (ECS) and engine anti-icing degrades SFC by 1–3%

**Engine types by mission:**
- Turbofan (BPR 5–15): long-range transports, regional jets
- Turboprop: short-haul, ISR, utility aircraft below ~400 kt
- Turboshaft: helicopters
- Turbojet/low-BPR turbofan: supersonic aircraft (Concorde, F/A-18)
- Electric motor + battery/fuel cell: emerging for short-range UAV and eVTOL

### Systems integration

"Systems" encompasses everything that is not structure, aerodynamics, or propulsion: hydraulics, avionics, flight control actuation, electrical power, environmental control, landing gear, fuel system, and ice protection.

Systems drive roughly 20–30% of aircraft empty weight and are responsible for a disproportionate share of development cost and schedule risk.
The A380 programme's famous 2006 delivery delay (causing a €4.8 billion write-down) was caused by wiring harness integration failures — a systems problem, not an aerodynamic or structural one.

```{warning}
Systems integration failures are the leading cause of schedule delays and cost overruns in modern aircraft programmes.
The discipline is complex because systems span the entire aircraft, interface with every other discipline, and have thousands of components — each with its own failure mode.
Model-Based Systems Engineering (MBSE) using SysML is increasingly used to manage this complexity from requirements through to detailed design.
```


### Flight performance

Performance analysis translates the aerodynamic polar and propulsion data into the mission outcomes that customers actually buy: range, speed, climb rate, ceiling, and field lengths.

Key performance equations:

| Metric | Governing equation |
|--------|-------------------|
| Range (jet) | $R = (V/\text{TSFC}) \cdot (L/D) \cdot \ln(W_i/W_f)$ |
| Climb rate | $\dot{h} = V(T - D)/W = V(T/W - D/W)$ |
| Take-off field | $s_{\text{TO}} \propto (W/S) / (\rho g C_{L_{\max}} (T/W))$ |
| Absolute ceiling | altitude where $T = D$ for the best-climb speed |

Performance is the discipline most directly experienced by the operator and therefore the one that drives the primary sizing trades.
Improving range by 3% — the equivalent of one extra row of seats flown "free" — can be worth hundreds of millions of dollars over a programme lifetime.

### Stability and control

A statically stable aircraft returns to trimmed flight after a disturbance.
A statically unstable aircraft requires active control (fly-by-wire) to remain flyable.

The degree of static stability is quantified by the static margin:

$$
K_n = \frac{x_{\text{np}} - x_{\text{cg}}}{\bar{c}}
$$

where $x_{\text{np}}$ is the neutral point (aerodynamic centre of the complete aircraft) and $x_{\text{cg}}$ is the centre of gravity, both expressed as fractions of the mean aerodynamic chord $\bar{c}$.

Civil transports typically fly with $K_n = 0.05$–$0.15$; relaxed-stability fighters (F-16, Eurofighter) fly with $K_n < 0$, relying on the flight control computer for stability augmentation.

### Manufacturing and producibility

A design that cannot be manufactured affordably is not a good design, regardless of its aerodynamic or structural merit.
Producibility considerations include:
- Minimum gauge constraints on composites and thin sheet metal
- Tooling access: can a rivet gun or drill reach the joint?
- Tolerance stack-up in large assemblies
- Interchangeability and replaceability of major components

### Economics and life-cycle cost

The aircraft that sells is not the one with the best aerodynamic performance but the one with the best economics for the operator's specific network.
Airline economics are driven by:

$$
\text{Cash Operating Cost} = f(\text{fuel burn, maintenance, crew, depreciation, fees})
$$

Fuel burn dominates for long-range aircraft ($\approx$ 30–40% of DOC); maintenance and depreciation dominate for short-haul operations.
A 1% improvement in fuel burn is worth approximately $\$500{,}000$–$\$2{,}000{,}000$ per aircraft over a 25-year service life, depending on the network {cite}`jenkinson2003`.

## Multidisciplinary design optimisation (MDO)

The coupling between disciplines makes classical sequential design — where each discipline optimises independently and then passes results to the next — inherently suboptimal.
In sequential design, the aerodynamicist optimises the wing for minimum drag, then hands the geometry to the structures team, who add weight to meet stress requirements, which degrades the aerodynamic performance — and the cycle repeats.

**Multidisciplinary Design Optimisation** (MDO) addresses this by solving all discipline analyses simultaneously, with a shared design vector and a global objective function {cite}`martins2021`.

The general MDO problem is:

$$
\min_{\mathbf{x}} f(\mathbf{x}, \mathbf{y})
\quad \text{subject to} \quad
\mathbf{g}(\mathbf{x}, \mathbf{y}) \leq 0, \quad
\mathbf{y} = \mathcal{A}(\mathbf{x}, \mathbf{y})
$$

where $\mathbf{x}$ is the design variable vector, $\mathbf{y}$ is the vector of coupled state variables (shared between disciplines), $f$ is the objective (e.g., fuel burn), and $\mathcal{A}$ represents the coupled analysis system.

**Common MDO architectures:**
- **MDF (Multidisciplinary Feasible):** Run the full coupled analysis to convergence at each optimiser iteration. Accurate but expensive.
- **IDF (Individual Discipline Feasible):** Each discipline optimises independently; consistency constraints enforce coupling. Parallelisable.
- **BLISS (Bi-Level Integrated System Synthesis):** Decomposed into system-level and discipline-level problems. Good for large systems.
- **CO (Collaborative Optimisation):** Suboptimise each discipline, then resolve inconsistencies at system level. Used at Boeing on the Blended Wing Body studies.

### Integrated Product Teams (IPTs)

In industry, MDO's organisational equivalent is the Integrated Product Team (IPT): a cross-disciplinary team that owns a component or system from requirements through manufacturing.
IPTs replaced the old sequential "throw it over the wall" model and became standard practice after the DoD's 1988 IPPD (Integrated Product and Process Development) directive.

The Boeing 777 "Working Together" programme organised teams by major structural component (wing, fuselage, empennage), with representatives from aerodynamics, structures, systems, and manufacturing co-located on the same floor — a physical manifestation of the coupling matrix.

### Digital twins and MBSE

Modern programmes increasingly maintain a live **digital twin** — a simulation model that is continuously updated with manufacturing and flight-test data and used to support in-service decisions, maintenance scheduling, and fleet management.

**Model-Based Systems Engineering (MBSE)** replaces document-centric requirements with a formal SysML model that captures requirements, interfaces, and behaviour in a machine-readable format.
This allows automatic consistency checking: if a requirements change propagates downstream, the model can flag every affected interface specification automatically.

## Case study: Airbus A380 vs. Boeing 787

These two programmes, launched within two years of each other, illustrate two opposite bets on the future of air travel — and two very different approaches to interdisciplinary integration.

| Attribute | Airbus A380 | Boeing 787 |
|-----------|------------|------------|
| Strategy | Hub-and-spoke: few very large aircraft between mega-hubs | Point-to-point: many medium aircraft on thinner routes |
| Passengers (3-class) | 555 | 242–330 |
| Range | 8,200 nmi | 7,355–8,500 nmi |
| Cruise Mach | 0.85 | 0.85 |
| Structure | Al-Li, GLARE, limited CFRP | 50% CFRP by weight |
| Engine | RR Trent 970, GE GP7200 | RR Trent 1000, GE GEnX |
| Fuel burn / seat | Baseline | ≈20% better |
| Development cost | ~€25 billion | ~$32 billion (with major overruns) |
| Key technical challenge | Four-engine integration, stairs, cabin systems | All-composite fuselage manufacturing, electrical architecture |
| Programme outcome | 251 aircraft delivered; production ended 2021 | 1,100+ deliveries; production continues |

The A380 proved the airport hub model was less prevalent than Airbus expected.
The 787 won the market bet but suffered severe development delays (3.5 years) due to the challenges of transitioning to large-scale composite fuselage manufacturing — an interdisciplinary problem spanning materials science, manufacturing process design, and supply chain management.
Both programmes reinforced the lesson that interdisciplinary complexity, managed poorly, is the primary driver of programme overruns {cite}`norris2010`.
