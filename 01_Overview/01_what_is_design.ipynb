# 1.1 — What is Aircraft Design?

Aircraft design begins with an explicitly identified **need**. That need ultimately limits and shapes the entire solution space — every design decision flows from it.

Three foundational definitions from leading authorities capture the scope of the discipline:

```{epigraph}
The intellectual engineering process of creating on paper a flying machine that either meets certain requirements and performance objectives, or explores new concepts, technologies, and innovations.

-- Anderson, *Introduction to Flight* {cite}`anderson2012`
```

```{epigraph}
The creation of the geometric description of something to be built.

-- Raymer, *Aircraft Design: A Conceptual Approach* {cite}`raymer2018`
```

```{epigraph}
About creating something with a purpose.

-- McGovern
```

What these definitions share is the idea that design is a **structured decision-making process under constraint**. The engineer must synthesise knowledge from aerodynamics, structures, propulsion, systems, and economics into a single coherent, buildable object.

Crucially, unlike a textbook problem with a unique correct answer, aircraft design admits a **multitude of valid solutions**. Two teams given identical requirements will produce different aircraft, both of which may be perfectly adequate. There is no 'best' aircraft in the abstract — only the best aircraft *for a specific mission, under specific constraints, at a specific point in time* {cite}`raymer2018`.

## A Brief History of Aircraft Design

The Wright Brothers solved a problem no one had solved before — and they solved it methodically. While their 1903 success is often portrayed as inspired tinkering, their engineering notebooks reveal careful hypothesis formation, systematic wind-tunnel testing of over 200 airfoil profiles, and deliberate documentation of results {cite}`anderson2012`. Their process was informal by modern standards, but it was recognisably *engineering*.

The decades that followed institutionalised that process. The National Advisory Committee for Aeronautics (NACA), founded in 1915, systematised aerodynamic knowledge — documenting airfoil families, compiling experimental data, and making it freely available to designers worldwide. By the 1930s and 1940s, a working engineer had reference books, not just intuition, to draw on.

World War II compressed what might have been decades of progress into years. The pressure to field superior aircraft faster than the adversary drove the development of systematic weight estimation methods, analytical performance prediction, and the first structured design review processes. The jet engine, which emerged from both Allied and Axis programmes simultaneously, fundamentally changed the performance envelope and forced the integration of entirely new propulsion disciplines into the design process.

The computational era began incrementally in the 1960s and accelerated through the 1970s and 1980s: finite element structural analysis displaced hand calculation, computational fluid dynamics began supplementing wind tunnels, and computer-aided design replaced the drawing board. By the 1990s, digital mock-up had made it possible to detect interference between subsystems before metal was cut — a change that fundamentally reduced development risk and cost {cite}`raymer2018`.

Today's practitioner has capabilities the NACA engineers could not have imagined: high-fidelity CFD, multidisciplinary optimisation frameworks, digital twins, and additive manufacturing. Yet the discipline's intellectual core — identify the need, define requirements, size the concept, iterate until it is right — has not changed since Orville and Wilbur worked in their bicycle shop in Dayton. The tools are different; the thinking is the same.

## Design vs. Drawing

A useful way to sharpen the definition is to contrast *engineering design* with what is often mistaken for it.

A common beginner workflow looks like this: a rough idea or a sketch is formed, it moves straight into CAD or construction, and a test flight follows — succeeding or failing largely by chance. There is no engineering between idea and build. This produces aircraft that may fly, but for reasons that are not well understood and cannot be systematically improved {cite}`gabefpv2026`.

A structured engineering design process inserts rigorous analysis at every step. Requirements are defined before any geometry is drawn. The aircraft is sized analytically. Performance is predicted from first principles. The flight test then becomes a *verification* of prior engineering rather than a roll of the dice. The difference is not just in quality of outcome — it is in understanding *why* the outcome happened, and being able to reproduce and improve it.

This distinction is especially visible at the small UAV scale, where the temptation to jump straight to building is highest. The engineering approach applies regardless of aircraft size: a 2 kg UAV and a 200-tonne transport both begin with requirements, not geometry {cite}`gabefpv2026` {cite}`raymer2018`.

````python
import matplotlib as mpl
mpl.rcParams.update({
    'figure.facecolor':      'white',
    'axes.facecolor':        'white',
    'axes.edgecolor':        '#aaaaaa',
    'axes.labelcolor':       '#333333',
    'text.color':            '#333333',
    'xtick.color':           '#333333',
    'ytick.color':           '#333333',
    'grid.color':            '#cccccc',
    'font.size':             11,
    'axes.titlesize':        13,
    'axes.labelsize':        11,
    'xtick.labelsize':       10,
    'ytick.labelsize':       10,
    'legend.fontsize':       10,
    'legend.title_fontsize': 10,
    'figure.titlesize':      13,
    'lines.linewidth':       2.0,
    'savefig.facecolor':     'white',
})
# Okabe-Ito colorblind-safe palette
OKI = ["#0072B2", "#D55E00", "#009E73", "#E69F00",
       "#56B4E9", "#CC79A7", "#F0E442", "#333333"]
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# Six clearly distinct node colors (dark enough for white text)
node_colors = ["#0072B2", "#b34000", "#006b40", "#8a5500", "#6b2060", "#1a6680"]
edge_colors = ["#56B4E9", "#e07040", "#30aa70", "#c8a030", "#b065a0", "#40a0b0"]

fig, ax = plt.subplots(figsize=(8, 8))
fig.patch.set_facecolor('white')
ax.set_facecolor('white')
ax.set_xlim(-1.65, 1.65)
ax.set_ylim(-1.65, 1.65)
ax.set_aspect('equal')
ax.axis('off')

disciplines = [
    'Aerodynamics',
    'Structures &\nMaterials',
    'Propulsion',
    'Systems\nIntegration',
    'Manufacturing &\nProducibility',
    'Economics &\nLifecycle Cost',
]

n = len(disciplines)
angles = np.linspace(np.pi / 2, np.pi / 2 + 2 * np.pi, n + 1)[:-1]
r_nodes = 1.15
positions = [(r_nodes * np.cos(a), r_nodes * np.sin(a)) for a in angles]

# Inter-node connection lines
for i in range(n):
    for j in range(i + 1, n):
        x1, y1 = positions[i]
        x2, y2 = positions[j]
        ax.plot([x1, x2], [y1, y2], color='#aaaaaa', alpha=0.25, lw=1.2, zorder=1)

# Centre hub
r_hub = 0.30
hub = plt.Circle((0, 0), r_hub, facecolor='#0072B2', edgecolor='#56B4E9', lw=2.2, zorder=5)
ax.add_patch(hub)
ax.text(0, 0, 'Aircraft\nDesign', fontsize=10, ha='center', va='center',
        color='white', fontweight='bold', multialignment='center', zorder=6)

# Spokes from hub to nodes
bw, bh = 0.60, 0.38
for i, (nx, ny) in enumerate(positions):
    dx, dy = nx, ny
    dist = np.hypot(dx, dy)
    ux, uy = dx / dist, dy / dist
    sx, sy = ux * r_hub, uy * r_hub
    tx = abs(ux); ty = abs(uy)
    if tx > 1e-9 and ty > 1e-9:
        t_box = min(bw / tx, bh / ty)
    elif tx < 1e-9:
        t_box = bh / ty
    else:
        t_box = bw / tx
    ex = nx - ux * (t_box + 0.04); ey = ny - uy * (t_box + 0.04)
    ax.annotate('',
        xy=(ex, ey), xytext=(sx, sy),
        arrowprops=dict(arrowstyle='->', color=edge_colors[i],
                        lw=2.0, mutation_scale=16),
        zorder=3)

# Discipline node boxes
for i, ((nx, ny), lbl) in enumerate(zip(positions, disciplines)):
    box = mpatches.FancyBboxPatch(
        (nx - bw, ny - bh), 2 * bw, 2 * bh,
        boxstyle='round,pad=0.10',
        facecolor=node_colors[i], edgecolor=edge_colors[i], lw=2.0, zorder=4)
    ax.add_patch(box)
    ax.text(nx, ny, lbl, fontsize=10, ha='center', va='center',
            color='white', fontweight='bold', multialignment='center', zorder=5)

plt.tight_layout(pad=0.2)
plt.show()
````

## The Breadth of the Discipline

Aircraft design does not fit inside any single engineering speciality. A complete design requires contributions from — and trade-offs between — all six of the disciplines shown in the figure above:

**Aerodynamics** shapes the outer surface to generate the required lift with minimum drag and ensures the aircraft remains controllable and stable throughout its flight envelope. Aerodynamic analysis provides the performance boundaries within which every other discipline must operate.

**Structures and materials** carry aerodynamic and inertial loads with the minimum possible mass. Structural decisions — spar location, material choice, skin gauge, joint design — ripple through to weight, stiffness, and ultimately to every performance metric. Nothing flies that cannot bear its own loads.

**Propulsion** provides the thrust or power to fly. Engine selection drives fuel capacity, installation geometry, cooling requirements, and noise footprint. For electric UAVs, the equivalent trade is between battery energy density, motor efficiency, and thermal management — but the core question is the same: how much energy does this mission require, and how is it stored and delivered?

**Systems integration** encompasses everything else onboard: avionics, flight control, landing gear, fuel distribution, hydraulics, electrical power, ice protection, and crew or operator interfaces. In complex aircraft, systems integration accounts for a substantial fraction of both development cost and airframe weight, and it is the discipline most frequently underestimated in early conceptual work.

**Manufacturing and producibility** constrain what can be built at the required cost and within the available schedule. The most aerodynamically optimal shape is worthless if it cannot be fabricated with existing tooling or processes. Design for manufacture is not a downstream activity — producibility constraints must influence geometry decisions from the earliest conceptual stage {cite}`raymer2018`.

**Economics and lifecycle cost** determine whether the aircraft can be developed, procured, operated, and maintained within the customer's budget. A technically superior design that is unaffordable is not a viable solution. Lifecycle cost analysis considers not just unit production cost but also fuel burn, maintenance labour, spares consumption, and eventual disposal.

These disciplines do not optimise in the same direction. The aerodynamicist wants a thin, high-aspect-ratio wing; the structures engineer wants a deep, heavily built one; the systems integrator wants fuel volume inside it; and the manufacturing engineer wants simple flat surfaces to fabricate. Aircraft design is the art of navigating these competing optima to find a single configuration that is *acceptable* across all of them simultaneously — not perfect in any one, but good enough in all of them together {cite}`nicolai2002`.

## Types of Aircraft Design Problems

Not all aircraft design work starts from a blank page. Practitioners recognise a spectrum of problem types, each with different degrees of freedom and different analytical approaches:

**New design (clean-sheet)** — the designer specifies the configuration, sizing, and all major parameters from scratch. Maximum freedom, maximum risk, and the longest development timeline. Clean-sheet programmes are rare because of their cost and schedule risk; when they occur, they typically follow a step-change in enabling technology or a genuinely new operational requirement that cannot be satisfied by any existing airframe. Examples span the full size range, from Scaled Composites' Voyager to the Airbus A380.

**Derivative design** — an existing aircraft is scaled, stretched, re-engined, or modified to serve a related mission. The Boeing 737 family has gone through eight successive variants across more than sixty years of continuous derivative development. Derivative design leverages known aerodynamics, established certification data, and existing supply chains — at the cost of inheriting the geometry and performance limitations of the original configuration.

**Modification and conversion** — an in-service aircraft is adapted for a new role without changing its basic type certificate. Tanker conversions, sensor-fitted maritime patrol variants, and special mission derivatives are common examples. The designer works within a tightly constrained envelope but benefits from a mature structural and systems baseline that substantially reduces analytical and certification risk.

**Paper study** — a design exercise conducted to answer a specific analytical question rather than to produce a buildable aircraft. Does a technology concept close on the stated requirements? What is the sensitivity of fuel burn to wing loading? These studies are the most common form of design work in research and advanced-concepts environments, and they are the environment in which the methods in this text are most directly applied.

In this guide, the focus is on **new design methodology** — the principles used to establish a clean-sheet design from requirements through to a credible conceptual configuration. These same principles apply, with appropriate adaptation, to derivative and modification work {cite}`raymer2018`.

## Design Objectives and Constraints

Every aircraft design problem has a precise mathematical structure, even when it is not stated explicitly. Understanding it prevents a common class of errors in early-stage analysis.

A **design objective** is what the design is optimised to achieve — a scalar quantity to be minimised or maximised. Formally, only one quantity can be the true objective at a time: minimum gross weight, maximum endurance, minimum unit cost, and minimum development time are each valid single objectives. Choosing the wrong objective — or failing to choose one — leads to designs that perform adequately on all metrics but are optimal on none.

A **design constraint** is a boundary the design must not violate. Constraints are stated as inequalities or equalities: takeoff distance ≤ 300 m, stall speed ≤ 18 m/s, gross weight ≤ 25 kg, structural safety factor ≥ 1.5. The **feasible design space** is the set of all configurations that satisfy every constraint simultaneously. The optimal design is the configuration that achieves the best objective value within that space.

Several failure modes arise when this structure is not observed clearly:

**Treating constraints as objectives.** "We need to minimise weight *and* maximise endurance *and* minimise cost" is not a well-posed optimisation. It requires either a priority ordering or an explicit weighting of objectives, both of which must be defined by the customer before design can proceed. An unresolved multi-objective statement generates paralysis rather than a design.

**Under-constrained problem specification.** A performance requirement stated without a bound ("long range") leaves the feasible space unbounded and makes analytical sizing impossible. Every requirement must have a number attached to it before the engineer can do useful work.

**Infeasible constraint sets.** A set of requirements that cannot simultaneously be satisfied — maximum cruise speed *and* minimum wingspan *and* minimum gross weight, all independently minimised — is a common early warning sign. Identifying infeasibility analytically, before any hardware or tooling is committed, is one of the most valuable contributions a conceptual designer can make to a programme {cite}`raymer2018` {cite}`nicolai2002`.

```{tip}
When beginning a new design, write down the objective explicitly — one line, one metric. Separate it clearly from the constraints. If a customer lists multiple objectives, ask: *if you had to choose one criterion by which this aircraft will be judged as a success or failure, what is it?* That answer is the true objective. Everything else becomes a constraint or a trade study.
```
