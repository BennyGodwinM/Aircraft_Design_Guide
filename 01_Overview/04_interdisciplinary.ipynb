# 1.4 — The Interdisciplinary Nature of Aircraft Design

Aircraft design cannot be decomposed into independent sub-problems. Every major discipline is tightly coupled to every other {cite}`torenbeek2013`. A change in aerodynamics propagates to structures, which propagates to weight, which propagates to propulsion, which propagates back to range and the mission requirements — restarting the design loop.

In industry, multidisciplinary design teams work in parallel with frequent integration reviews to keep disciplines synchronised. No single engineer 'solves' an aircraft — it emerges from the collective interaction of specialists {cite}`raymer2018`.

A key skill in aircraft design is therefore not just mastery of one discipline, but the ability to reason across discipline boundaries — to understand how a decision in one area ripples through the rest of the system.

```{note}
The disciplines that must be coordinated in any aircraft design include at minimum: aerodynamics, structural design, propulsion, flight performance, stability and control, and systems integration. Each feeds into the others, and none can be finalised in isolation.
```
