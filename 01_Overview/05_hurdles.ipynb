# 1.5 — The Hurdles of Aircraft Design

The following challenges are persistent across the industry and must be understood before serious design work begins.

**The Tyranny of Weight.** Every kilogram added requires more lift, which requires more wing area, which adds structural weight, which requires more thrust, which requires a heavier engine, which burns more fuel, which adds more weight. This compounding cycle — the *weight snowball* — means that every design decision has non-linear consequences. Minimising weight while maintaining structural integrity is a constant, unresolvable tension {cite}`raymer2018`.

**Competing Requirements.** A faster aircraft is usually less efficient. A longer-range aircraft is usually heavier. A cheaper aircraft is usually less capable. The designer must navigate this trade-space without the ability to satisfy all constraints simultaneously — a challenge that is as much a matter of negotiation as mathematics {cite}`torenbeek2013`.

**The Regulatory and Certification Burden.** Aircraft must be certified airworthy before commercial operation. The FAA (FARs) and EASA certification process is lengthy, expensive, and deeply constraining. Designing to be certifiable — not just functional — is a discipline in itself. A concept that cannot be certified is not a viable aircraft, regardless of its technical merits {cite}`raymer2018`.

**Long Timelines and Cost.** A modern commercial transport takes 10–15 years and tens of billions of dollars from concept to first delivery. Forecasting errors and technical setbacks compound over time, making cost overruns and schedule delays endemic rather than exceptional {cite}`torenbeek2013`.

**The Gap Between Analysis and Reality.** No matter how sophisticated the computational models, the aircraft must be built and flown. The gap between predicted and actual performance is a perennial source of surprises. Managing this uncertainty through conservative margins and staged testing is fundamental to professional practice {cite}`nicolai2002`.

**Iteration vs. Innovation.** The safest path is incremental improvement on a proven concept. Radical innovations carry unknown risks and face a sceptical regulatory system. Most promising ideas fail to reach production not because they don't work aerodynamically, but because they cannot be certified, manufactured at scale, or sustained economically {cite}`anderson2012`.

## Typical Mission Profile

The figure below shows a schematic mission profile for a civil transport aircraft. This profile — from taxi and takeoff through climb, cruise, descent, and landing — defines the performance requirements that drive the entire design {cite}`raymer2018` {cite}`torenbeek2013`.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import PchipInterpolator

plt.rcParams['savefig.facecolor'] = 'none'

fig, ax = plt.subplots(figsize=(10, 4.2))
fig.patch.set_facecolor('none')
ax.set_facecolor('none')

px = [0.0, 0.3, 0.6, 1.5, 2.2, 8.5, 9.3, 9.9, 10.4, 10.7]
py = [0.0, 0.0, 0.5, 3.2, 4.0, 4.0, 3.2, 0.5, 0.0, 0.0]
interp = PchipInterpolator(px, py)
xs = np.linspace(0, 10.7, 500)
ys = np.clip(interp(xs), 0, None)

ax.fill_between(xs, 0, ys, color='#1565c0', alpha=0.10, linewidth=0, zorder=1)
ax.plot(xs, ys, color='#1565c0', lw=2.5, zorder=4)
ax.axhline(0, color='#667788', lw=1.0, zorder=3)

for xd in [0.6, 2.2, 8.5, 9.3, 9.9]:
 ax.axvline(xd, color='#99aabb', lw=0.9, linestyle='--', alpha=0.5, zorder=2)

phase_defs = [
 (0.0, 0.6, 'Taxi &\nTakeoff', 0.25),
 (0.6, 2.2, 'Climb', 1.4),
 (2.2, 8.5, 'Cruise', 5.35),
 (8.5, 9.3, 'Descent', 8.9),
 (9.3, 9.9, 'Approach &\nLanding', 9.6),
 (9.9, 10.7, 'Taxi', 10.3),
]
for x0, x1, lbl, lx in phase_defs:
 mid_y = float(interp((x0 + x1) / 2))
 label_y = min(max(mid_y + 0.25, 0.5), 3.6)
 ax.text(lx, label_y + 0.45, lbl,
 fontsize=8.5, ha='center', va='bottom',
 color='#1a2c42', multialignment='center',
 bbox=dict(boxstyle='round,pad=0.3', facecolor='none',
 edgecolor='#1565c0', lw=1.1),
 zorder=5)
 ax.annotate('', xy=(lx, label_y + 0.04), xytext=(lx, label_y + 0.42),
 arrowprops=dict(arrowstyle='->', color='#1565c0', lw=1.0),
 zorder=5)

for y_val, lbl in [(0, '0'), (2.0, '~5,000 ft'), (4.0, '~35,000 ft')]:
 ax.axhline(y_val, color='#99aabb', lw=0.5, linestyle=':', alpha=0.5)
 ax.text(-0.2, y_val, lbl, fontsize=8, ha='right', va='center', color='#445566')

ax.set_xlim(-0.4, 11.0)
ax.set_ylim(-0.4, 5.4)
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('Mission Progress →', fontsize=10, color='#334455', labelpad=4)
ax.spines[['top', 'right', 'left', 'bottom']].set_visible(False)

plt.tight_layout()
plt.show()
````
