# 11.2 — A History of Aerospace Propulsion

The history of aerospace propulsion is a story of relentless pursuit: more power from less weight, more thrust from less fuel, more speed from less noise. Each generation of engineers inherited unsolved problems from the last and created new ones for the next. The transition from a 12-horsepower piston engine in 1903 to a turbofan producing 115,000 pounds of thrust in 2010 represents one of the most compressed technological trajectories in human history.

This notebook traces that trajectory — from the earliest dreams of powered flight through the propeller era, the jet revolution, the turbofan age, and the emerging era of electric and sustainable propulsion.

## 1. The Pre-Flight Era: Steam, Clockwork, and Dreams

### Ader, Langley, and the Steam Epoch (1890–1903)

Before the internal combustion engine reached sufficient power-to-weight ratios, early aviation pioneers were forced to consider steam. The fundamental problem was brutal: a steam engine producing 10 kW might weigh 200 kg or more — a power-to-weight ratio of 0.05 kW/kg, compared to the 0.5–1.0 kW/kg needed for sustained flight.

**Clément Ader** (1841–1925) built the *Éole* (1890) and *Avion III* (1897), both powered by lightweight steam engines of his own design. The *Avion III* used a 15 kW engine weighing just 50 kg — remarkable for its era. French Army tests in 1897 produced contested evidence of short hops, but no sustained controlled flight was achieved.

**Samuel Langley** (1834–1906), Secretary of the Smithsonian Institution, developed the *Aerodrome A* powered by a 39 kW radial gasoline engine designed by Charles Manly — achieving a power-to-weight ratio of 0.47 kW/kg, far superior to any steam alternative. The *Aerodrome* failed twice on launch catapult malfunction (December 1903), just nine days before Kitty Hawk. The engine itself was technically superior to the one the Wrights used; the airframe was not.

### The Internal Combustion Engine Arrives

The four-stroke gasoline engine, patented by Nikolaus Otto in 1876, offered a fundamentally better power-to-weight ratio than steam. By 1900, automotive engines had demonstrated that 15–30 kW outputs were achievable at weights under 100 kg. The aviation market, however, required higher power-to-weight ratios, lower vibration (for structural integrity), and reliable spark ignition — requirements that pushed engine manufacturers to develop dedicated aviation powerplants.

## 2. The Wright Engine and the Dawn of Powered Flight (1903)

The Wright brothers were exceptional engineers who recognised that no off-the-shelf automobile engine of 1902 met their requirements — too heavy, too unreliable, and carburation problems at low temperatures. With their mechanic **Charles Taylor**, they designed and built their own engine in six weeks.

### The 1903 Wright Engine — Specifications

| Parameter | Value |
|-----------|-------|
| Configuration | 4-cylinder, horizontal, water-cooled |
| Displacement | 3.29 litres |
| Power output | ~8.9 kW (12 hp) at 1,025 rpm |
| Weight (dry) | 82 kg |
| Power-to-weight | 0.11 kW/kg |
| Ignition | Make-and-break contact (no throttle) |
| Fuel | Gasoline |
| Drive | Chain drive to two counter-rotating pusher propellers |

The engine had no throttle, no carburetor in the conventional sense (fuel dripped into a heated manifold), and no oil pump — it relied on a gravity-fed splash lubrication system. It ran for only 12 seconds on the first flight. Its genius lay not in sophistication but in being *just good enough* — combined with the brothers' exceptionally efficient 8.5-foot propellers (achieving ~66% efficiency) and their light, strong airframe.

### The Propeller — An Underrated Innovation

The Wrights' analytical approach to propeller design was as significant as their engine. Most contemporaries designed propellers by analogy with marine screws. The Wrights recognised that an aircraft propeller is a **rotating wing** and applied their wind-tunnel aerofoil data to blade design. Their propellers achieved 66% efficiency — comparable to modern wooden propellers and far superior to the 40–50% of their contemporaries.

## 3. World War I — The Rotary Engine and the Race for Power (1914–1918)

### The Rotary Engine

The dominant fighter engine of 1914–1917 was the **rotary** — a configuration in which the entire engine rotated around a stationary crankshaft, with the propeller bolted to the crankcase. The Gnome Monosoupape (80 hp), Le Rhône 9J (110 hp), and Clerget 9B (130 hp) powered the Sopwith Camel, Nieuport 17, and virtually every Allied fighter of the mid-war period.

The rotary's genius was its power-to-weight ratio (~0.45 kW/kg) and cooling efficiency (the spinning cylinders cooled themselves in the airstream). Its limitations were severe:
- **Gyroscopic precession**: The spinning engine mass created powerful gyroscopic moments that made aircraft extremely responsive to control in one direction and sluggish in the other — the Sopwith Camel's notorious left-turn tendency killed many student pilots.
- **Throttle control**: The castor oil lubricant mixed with fuel and could not be recycled; the only throttle was a 'blip switch' that cut ignition entirely.
- **Power ceiling**: Limited to ~150 hp by cooling and structural constraints.

### The Stationary In-Line Engine — The Path Forward

By 1917, the rotary was giving way to the **water-cooled V-8 and V-12 in-line engines** — the Hispano-Suiza 8 (150–220 hp), the Rolls-Royce Eagle (360 hp), and the Liberty L-12 (400 hp). These engines permitted much higher power outputs through liquid cooling and multi-cylinder configurations. The Liberty L-12, designed in 5 days and mass-produced in 6 months in 1917, was a landmark in industrial-scale aviation engine production.

By the Armistice, fighter engines had progressed from 80 hp to 400+ hp — a 5× increase in four years, with power-to-weight ratios improving from 0.4 to over 0.7 kW/kg.

## 4. The Interwar Period — Superchargers, Racing, and the Merlin (1919–1939)

### The Schneider Trophy and Racing Engines

The Schneider Trophy races (1913–1931) for seaplanes were the Formula 1 of aviation, and their engines were the crucible in which the Spitfire's Merlin was forged. The winning Supermarine S.6B (1931, 406 mph) was powered by an experimental Rolls-Royce R engine producing 2,350 hp from 37 litres — a power density of 63 hp/litre, compared to ~15 hp/litre for contemporary production engines.

The lessons of the R engine were directly transferred to the **Rolls-Royce Merlin** (1936), which produced 1,030 hp in early mark and eventually 2,070 hp in the Merlin 130/131, through:
- Two-speed, two-stage supercharging
- 100-octane fuel (enabling higher boost pressures)
- Better intercooling
- Continuous improvement of valve timing and combustion chamber geometry

### The Piston Engine's Pinnacle

By 1944, the piston aero-engine had reached a point of extraordinary sophistication — and near its absolute limit. The **Rolls-Royce Griffon** (2,340 hp), **Napier Sabre** (2,400 hp, 24-cylinder H-configuration), and **Pratt & Whitney R-4360 Wasp Major** (3,500 hp, 28-cylinder, 4-row radial) represented the maximum achievable with spark-ignition reciprocating technology.

The fundamental constraint was the propeller: above approximately 750 km/h, compressibility effects at the blade tips caused efficiency collapse and structural vibration. No further speed increase was possible without abandoning the propeller entirely.

## 5. The Jet Revolution — Whittle and von Ohain (1930–1945)

### Frank Whittle — The Theoretical Breakthrough

**Frank Whittle** (1907–1996), a Royal Air Force cadet, submitted his thesis *Future Developments in Aircraft Design* in 1928, proposing gas turbine propulsion. He filed his first jet engine patent in January 1930 — a year before the RAF showed any interest. Working with private funding and near-constant technical setbacks, Whittle's company Power Jets Ltd ran the first successful turbojet bench test on 12 April 1937.

The engine — the **W.1** — produced 3.8 kN of thrust and first flew in the Gloster E.28/39 on 15 May 1941. Its pressure ratio was only 4:1 (modern engines use 40–60:1), and its thrust-specific fuel consumption was nearly double that of a good piston engine, but the principle was proven.

### Hans von Ohain — Independent Discovery

Unknown to Whittle, **Hans von Ohain** (1911–1998) independently arrived at the jet propulsion concept in Germany. Working with Ernst Heinkel's private aircraft company, von Ohain's **HeS 3** turbojet (4.4 kN thrust) powered the **Heinkel He 178** on 27 August 1939 — the world's first turbojet-powered flight, three days before Germany invaded Poland.

### The Me 262 — First Jet in Combat

The **Messerschmitt Me 262** (1944) was the first operational jet-powered fighter, powered by two Junkers Jumo 004B turbojets (8.8 kN each). Its 870 km/h top speed was 180 km/h faster than any Allied piston fighter. The Jumo 004 was a landmark engine: the first turbojet to enter mass production, though its 25-hour overhaul life (vs. 1,000 hours for a Merlin) reflected the desperate scarcity of chromium and nickel in wartime Germany.

### The Gloster Meteor — Allied Response

The **Gloster Meteor** entered RAF service in July 1944, becoming the only Allied jet to see operational service in WWII. Powered by two Rolls-Royce Welland engines (derived from Whittle's W.2), it demonstrated that the British had maintained parallel development despite sharing technology with the Americans.

## 6. The Cold War and the Turbojet Era (1945–1965)

### Swept Wings and the Sound Barrier

Post-war analysis of captured German research (particularly Busemann's swept-wing compressibility data from the AVA Göttingen) directly shaped the first generation of Cold War jets. The **North American F-86 Sabre** (1947) and **MiG-15** (1947) — the protagonists of the Korean War — both used 35° swept wings adapted from German data.

The **Bell X-1**, powered by a four-chamber XLR11 rocket engine (26.7 kN total), became the first aircraft to break the sound barrier on 14 October 1947, with Chuck Yeager at the controls at Mach 1.06. The X-1's engine was a rocket because jet engines of 1947 could not sustain supersonic flight — too little thrust, too little fuel efficiency.

### First-Generation Turbojets

The first decade of jet aviation was characterised by rapid improvement in basic turbojet technology. Pressure ratios rose from 4:1 (Whittle W.1) to 12:1 (Rolls-Royce Avon, 1950) to 20:1 (Pratt & Whitney J57, 1951). The J57 — the engine in the F-100 Super Sabre and B-52 — was the first production turbojet to exceed 10,000 lbf thrust and the first with a two-spool (twin-shaft) architecture, which allowed the high-pressure and low-pressure compressors to rotate at their optimal speeds independently.

### The Commercial Jet Age Begins

The **de Havilland Comet** (1952) entered service with four de Havilland Ghost turbojets buried in the wing roots. Its catastrophic pressure-cabin fatigue failures (1954) grounded the fleet and gave Boeing the time to perfect the **707** (1958), powered by four Pratt & Whitney JT3C/D turbojets (later converted to turbofans), which established the template for every subsequent jet airliner.

## 7. The Turbofan Revolution (1960–1990)

### The Bypass Ratio Insight

The propulsive efficiency penalty of the pure turbojet at subsonic cruise was well understood by the late 1950s. The solution — a fan driven by an additional turbine stage to move a bypass airstream around the core — had been proposed by A. A. Griffith at the RAE as early as 1926 and patented by Whittle in 1936. The question was not *whether* to use bypass but *how much*.

The first turbofans were **low-bypass** (BPR 0.3–1.5), introduced primarily to reduce specific fuel consumption for long-range bombers:
- **Pratt & Whitney JT3D** (1960, BPR 1.4): Converted the JT3C turbojet by adding a two-stage fan and extra turbine stages. The JT3D reduced TSFC by 25% and became the standard engine for the Boeing 707-120B, 707-320B, and DC-8.
- **Rolls-Royce Conway** (1960, BPR 0.3): The first bypass engine to enter commercial service, powering the Vickers VC10 and Boeing 707-420.

### The High-Bypass Breakthrough — JT9D and the 747

The step change to **high-bypass turbofans** was driven by the Boeing 747 programme. Boeing required an engine capable of 45,000 lbf thrust — nearly twice anything previously flown — to power the 350-seat wide-body. Pratt & Whitney's **JT9D** (1969, BPR 4.8) was the result: a dramatic departure from existing engine sizes, with a 2.4 m fan diameter. The concurrent development of the **General Electric TF39** (BPR 8.0) for the Lockheed C-5 Galaxy demonstrated that even higher bypass ratios were feasible and beneficial.

The high-bypass turbofan delivered:
- 25–30% reduction in TSFC versus equivalent thrust turbojet
- Dramatic noise reduction (lower jet velocity = lower exhaust noise)
- Improved hot-section life (lower gas velocities, better cooling)

### The CFM56 — Greatest Commercial Engine Programme

The **CFM International CFM56** (1974), a joint venture between GE and SNECMA, became the best-selling commercial turbofan engine in history with over 33,000 units delivered. Its BPR of 5–6 and TSFC of ~0.55 lb/(lbf·h) set the efficiency benchmark for the 737 Classic and 737NG, A320, and military tankers for forty years.

## 8. Supersonic Propulsion — Afterburners and Concorde (1950–2003)

### Military Augmentation

The Korean War demonstrated that jet fighters needed transient thrust far beyond what was sustainable on fuel efficiency grounds. The **afterburner** (reheat) — additional fuel injected into the turbine exhaust and burned in a long combustion tube aft of the turbine — was the answer. First flown operationally in the J35 Allison (1948), it became universal on combat aircraft by the mid-1950s.

Afterburner engineering challenges include:
- **Flameholding**: Supersonic nozzle exit flow must be stabilised for combustion. V-gutter flameholders create recirculation zones.
- **Variable nozzle**: The greatly increased gas volume requires an expanding nozzle area, controlled by iris-type or convergent-divergent variable geometry.
- **Thermal management**: Liner temperatures exceed 1,800 K; transpiration cooling and exotic ceramics are essential.

### The Concorde Olympus 593

The **Rolls-Royce/SNECMA Olympus 593** powering Concorde represented a unique engineering achievement: a pure turbojet (no bypass — a bypass stream would overheat at M 2.04) with reheat, producing 169.3 kN (38,000 lbf) with afterburner for take-off, and cruising at 169.3 kN dry at M 2.04 / 55,000 ft — where the ram pressure ratio of the inlet alone provided a compression ratio of approximately 7.3.

## 9. The Ultra-High Bypass Era (1990–Present)

### Pushing Bypass Ratios Past 9

By the 1990s, advances in computational fluid dynamics, single-crystal turbine blade technology, and improved thermal barrier coatings allowed engine manufacturers to push bypass ratios beyond 8–9 while simultaneously increasing overall pressure ratios above 40. The dual imperatives of fuel economy (driven by fuel price) and noise reduction (driven by increasingly strict airport regulations) drove this trend.

**Key engines of the 1990s–2000s:**
- **GE90-115B** (1995, uprated 2003): BPR 9.0, OPR 42, 115,540 lbf — the world's most powerful commercial turbofan, powering the Boeing 777-300ER. First commercial engine with full carbon-fibre composite fan blades.
- **Rolls-Royce Trent 800** (1996, BPR 6.4) and **Trent XWB** (2013, BPR 9.3, OPR 52): Three-shaft architecture allows each spool to rotate at its optimum speed, an approach unique to Rolls-Royce.
- **GEnx** (2008, BPR 8.8): Powers Boeing 787 and 747-8; uses CMC turbine shrouds and improved film cooling.

### The Geared Turbofan — Pratt & Whitney PW1000G

The **Pratt & Whitney PW1100G** (2016, BPR 12.5) introduced the geared turbofan to the single-aisle market on the Airbus A320neo family. The epicyclic reduction gearbox (gear ratio 3:1) allows the fan to rotate at ~3,000 rpm while the low-pressure turbine turns at ~9,000 rpm — each at its optimal speed. The result: 15–20% lower fuel burn versus the engines (CFM56, V2500) it replaced, with 50% lower noise footprint.

### RISE and the Open-Fan Future

CFM International's **RISE** programme (Revolutionary Innovation for Sustainable Engines, announced 2021) targets a 20% further fuel burn reduction over the LEAP engine through an **open-fan** (unducted fan) architecture with variable-pitch composite blades. The open fan avoids the nacelle drag and weight of a conventional duct while achieving BPR > 70, dramatically improving propulsive efficiency. Entry into service is targeted for the late 2030s.

## 10. Electric, Hydrogen, and Sustainable Propulsion (2000–Present)

### Early Electric Flight

The first electrically powered crewed aircraft flights date to the 1970s (Militky/Brditschka *MB-E1*, 1973), but the technology was too limited for practical application. Modern electric aviation began in earnest with the **Solar Impulse** projects (Piccard & Borschberg): *Solar Impulse 2* completed a circumnavigation in 2016 on solar power alone, demonstrating sustained electric flight despite very low speed (< 90 km/h) and payload (1 pilot).

### NASA X-57 Maxwell and Distributed Electric Propulsion

NASA's **X-57 Maxwell** (development 2015–2022) demonstrated the distributed electric propulsion (DEP) concept: 14 small electric motors on the wing leading edge augment lift at low speed, enabling a smaller wing (reducing cruise drag) while maintaining safe take-off and landing speeds. Only two larger wingtip motors are used in cruise.

### Hydrogen Propulsion

Hydrogen offers 2.8× the specific energy of kerosene (33.3 MJ/kg vs. 11.9 MJ/kg by volume), zero carbon emissions, and no SOₓ. Its challenges — cryogenic storage at −253°C, low volumetric energy density (requires ~4× the tank volume of Jet-A for equivalent energy), embrittlement of metals, and complete absence of airport infrastructure — are formidable but not fundamentally unsolvable.

**Airbus ZEROe** (announced 2020) proposes three hydrogen aircraft concepts targeting entry into service by 2035:
1. A turbofan-powered aircraft using hydrogen combustion in a modified gas turbine (max ~200 seats, range ~2,000 nm)
2. A turboprop concept for regional routes (< 100 seats)
3. A blended-wing-body concept with embedded hydrogen propulsion

### Sustainable Aviation Fuels (SAF)

Drop-in **sustainable aviation fuels** — biofuels and synthetic fuels produced from renewable electricity (power-to-liquid) — can reduce lifecycle carbon emissions by 70–90% without modifying existing engines or infrastructure. Current SAF production covers less than 0.1% of global aviation fuel demand; scaling requires enormous agricultural/energy inputs and policy support. ICAO's CORSIA scheme targets carbon-neutral growth post-2019 levels, with SAF as the primary mechanism.

### The 2050 Landscape

The most credible net-zero pathway for aviation by 2050 involves a combination of:
- **Ultra-high-bypass / open-fan** engines for existing-range missions
- **SAF** for medium- and long-haul routes that cannot be electrified
- **Hydrogen gas turbines** for new medium-haul platforms (2035–2040)
- **Full electrification** for short-haul (< 500 km) commuter routes
- **eVTOL** for urban air mobility replacing short road trips

## 11. Recurring Themes in Propulsion History

Looking across 120 years of propulsion development, several themes recur:

**1. Military necessity as the primary driver.** Every major propulsion breakthrough — the WWI Liberty engine, the WWII Merlin, the Cold War turbojet, the afterburner — was funded and urgently needed by military programmes. Commercial development followed, exploiting the technology at lower cost once risks were understood.

**2. Materials technology as the ultimate bottleneck.** Engine performance is ultimately limited by what the hottest turbine blades can withstand. Every major efficiency step — from cast to directionally solidified to single-crystal turbine blades, from aluminium to titanium compressor discs, from metallic to CMC combustor liners — came from materials science, not thermodynamic insight.

**3. The efficiency-speed trade-off.** Every engine type occupies a speed regime where it is efficient; outside that regime it is wasteful. Turboprops dominate below M 0.7; turbofans from M 0.7 to 1.0; turbojets from M 1.5 to 3.0; ramjets above M 2.5. The future may belong to variable-cycle engines that morph between regimes.

**4. The role of government-funded research.** The NACA (later NASA), RAE, and their counterparts funded the fundamental aerodynamic and thermodynamic research that underpinned each generation. The compressor aerodynamics, high-temperature materials, and manufacturing techniques in every modern turbofan trace their lineage to publicly funded research programmes.

**5. Incremental improvement, punctuated by paradigm shifts.** The 55% reduction in TSFC since the first turbojets was achieved through continuous incremental improvement (better materials, higher OPR, better compressor efficiency) punctuated by occasional architectural shifts (adding bypass, adding gearboxes, adopting composite fans). The open-fan and hydrogen transitions under development may be the next such shift.
