# 27.4 — Novel & Unconventional Configurations

Examines aircraft configurations that depart significantly from the conventional tube-and-wing layout: blended wing body, flying wings, joined wings, box wings, and tailless designs. Each presents distinct aerodynamic advantages and stability/control challenges.

**Key topics to cover:**

- Passenger cabin layout and emergency egress in non-cylindrical fuselages
- Notable designs: Boeing X-48, Airbus MAVERIC
